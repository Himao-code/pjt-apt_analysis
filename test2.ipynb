```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# í•œê¸€ í°íŠ¸ ì„¤ì •
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False

class RealEstateAnalyzer:
    def __init__(self, data_path='data/'):
        self.data_path = data_path
        self.regions = ['ê´‘ì£¼', 'ë¶€ì‚°', 'ì œì£¼']
        self.raw_data = {}
        self.processed_data = {}
        
    def load_all_data(self):
        """ëª¨ë“  ì§€ì—­ì˜ ë°ì´í„°ë¥¼ ë¡œë“œ"""
        print("=== ë°ì´í„° ë¡œë”© ì‹œì‘ ===")
        
        for region in self.regions:
            self.raw_data[region] = []
            print(f"\n[{region}] ë°ì´í„° ë¡œë”© ì¤‘...")
            
            for i in range(1, 6):  # 1~5ë²ˆ íŒŒì¼
                try:
                    filename = f"{region}_ì•„íŒŒíŠ¸_{i}.csv"
                    filepath = f"{self.data_path}{filename}"
                    
                    # ë‹¤ì–‘í•œ ì¸ì½”ë”©ìœ¼ë¡œ ì‹œë„
                    for encoding in ['utf-8', 'cp949', 'euc-kr']:
                        try:
                            df = pd.read_csv(filepath, encoding=encoding)
                            print(f"  âœ“ {filename} ë¡œë“œ ì™„ë£Œ ({len(df)} rows, {len(df.columns)} cols)")
                            self.raw_data[region].append(df)
                            break
                        except UnicodeDecodeError:
                            continue
                    else:
                        print(f"  âœ— {filename} ì¸ì½”ë”© ì˜¤ë¥˜")
                        
                except Exception as e:
                    print(f"  âœ— {filename} ë¡œë“œ ì‹¤íŒ¨: {e}")
        
        return self.raw_data
    
    def explore_data_structure(self):
        """ë°ì´í„° êµ¬ì¡° íƒìƒ‰"""
        print("\n=== ë°ì´í„° êµ¬ì¡° ë¶„ì„ ===")
        
        for region in self.regions:
            print(f"\n[{region}] ë°ì´í„° êµ¬ì¡°:")
            
            if not self.raw_data[region]:
                print("  ë°ì´í„° ì—†ìŒ")
                continue
                
            for i, df in enumerate(self.raw_data[region], 1):
                print(f"  íŒŒì¼ {i}: {df.shape}")
                print(f"  ì»¬ëŸ¼: {list(df.columns)}")
                
                # ì²« ë²ˆì§¸ íŒŒì¼ì˜ ìƒ˜í”Œ ë°ì´í„° í‘œì‹œ
                if i == 1:
                    print(f"  ìƒ˜í”Œ ë°ì´í„°:")
                    print(df.head(3).to_string())
                print("-" * 50)
    
    def combine_region_data(self):
        """ì§€ì—­ë³„ ë°ì´í„° ê²°í•©"""
        print("\n=== ì§€ì—­ë³„ ë°ì´í„° ê²°í•© ===")
        
        for region in self.regions:
            if not self.raw_data[region]:
                continue
                
            try:
                # ëª¨ë“  íŒŒì¼ì„ í•˜ë‚˜ë¡œ ê²°í•©
                combined_df = pd.concat(self.raw_data[region], ignore_index=True)
                
                # ì§€ì—­ ì •ë³´ ì¶”ê°€
                combined_df['ì§€ì—­'] = region
                
                self.processed_data[region] = combined_df
                print(f"{region}: {len(combined_df)} ê±´ì˜ ë°ì´í„° ê²°í•© ì™„ë£Œ")
                
            except Exception as e:
                print(f"{region} ë°ì´í„° ê²°í•© ì‹¤íŒ¨: {e}")
    
    def basic_statistics(self):
        """ê¸°ë³¸ í†µê³„ ì •ë³´"""
        print("\n=== ê¸°ë³¸ í†µê³„ ì •ë³´ ===")
        
        for region in self.regions:
            if region not in self.processed_data:
                continue
                
            df = self.processed_data[region]
            print(f"\n[{region}]")
            print(f"ì´ ê±°ë˜ ê±´ìˆ˜: {len(df):,}ê±´")
            
            # ê°€ê²© ê´€ë ¨ ì»¬ëŸ¼ì´ ìˆëŠ”ì§€ í™•ì¸
            price_cols = [col for col in df.columns if any(keyword in col.lower() 
                         for keyword in ['ê°€ê²©', 'ê¸ˆì•¡', 'price', 'ê±°ë˜ê¸ˆì•¡', 'ë§¤ë§¤ê°€'])]
            
            if price_cols:
                for col in price_cols[:3]:  # ìƒìœ„ 3ê°œë§Œ í‘œì‹œ
                    if df[col].dtype in ['int64', 'float64']:
                        print(f"{col}:")
                        print(f"  í‰ê· : {df[col].mean():,.0f}")
                        print(f"  ì¤‘ì•™ê°’: {df[col].median():,.0f}")
                        print(f"  ìµœëŒ€ê°’: {df[col].max():,.0f}")
                        print(f"  ìµœì†Œê°’: {df[col].min():,.0f}")
    
    def run_initial_analysis(self):
        """ì´ˆê¸° ë¶„ì„ ì‹¤í–‰"""
        self.load_all_data()
        self.explore_data_structure() 
        self.combine_region_data()
        self.basic_statistics()
        
        return self.processed_data

# ë¶„ì„ ì‹¤í–‰
analyzer = RealEstateAnalyzer()
data = analyzer.run_initial_analysis()

print("\n=== ì´ˆê¸° ë¶„ì„ ì™„ë£Œ ===")
print("ë‹¤ìŒ ë‹¨ê³„ì—ì„œëŠ” ê°€ê²© ë³€ë™ íŒ¨í„´ê³¼ íˆ¬ì ë§¤ë ¥ë„ë¥¼ ë¶„ì„í•˜ê² ìŠµë‹ˆë‹¤.")
```

```python
# ì•„íŒŒíŠ¸ ì´ ê±°ë˜ëŸ‰ ë¶„ì„


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# í•œê¸€ í°íŠ¸ ì„¤ì •
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False

class StepByStepAnalyzer:
    def __init__(self, data_path='data/'):
        self.data_path = data_path
        self.regions = ['ê´‘ì£¼', 'ë¶€ì‚°', 'ì œì£¼']
        self.raw_data = {}
        
    def load_and_explore_raw_data(self):
        """ì›ë³¸ ë°ì´í„° ë¡œë“œ ë° êµ¬ì¡° íƒìƒ‰"""
        print("=== 1ë‹¨ê³„: ì›ë³¸ ë°ì´í„° êµ¬ì¡° íŒŒì•… ===")
        
        for region in self.regions:
            print(f"\nğŸ“ [{region}] ë°ì´í„° ë¶„ì„:")
            self.raw_data[region] = []
            
            for i in range(1, 6):
                try:
                    filename = f"{region}_ì•„íŒŒíŠ¸_{i}.csv"
                    filepath = f"{self.data_path}{filename}"
                    
                    # ë‹¤ì–‘í•œ ì¸ì½”ë”©ìœ¼ë¡œ ì‹œë„
                    for encoding in ['utf-8', 'cp949', 'euc-kr']:
                        try:
                            df = pd.read_csv(filepath, encoding=encoding)
                            self.raw_data[region].append(df)
                            
                            print(f"  ğŸ“„ {filename}: {len(df)}ê±´, {len(df.columns)}ê°œ ì»¬ëŸ¼")
                            
                            # ì²« ë²ˆì§¸ íŒŒì¼ì˜ ì»¬ëŸ¼ êµ¬ì¡° ìƒì„¸ ë¶„ì„
                            if i == 1:
                                print(f"  ğŸ“‹ ì»¬ëŸ¼ ëª©ë¡:")
                                for idx, col in enumerate(df.columns, 1):
                                    print(f"    {idx:2d}. {col}")
                                
                                print(f"  ğŸ“Š ìƒ˜í”Œ ë°ì´í„° (ìƒìœ„ 3ê±´):")
                                print(df.head(3).to_string())
                                print("-" * 80)
                            break
                        except UnicodeDecodeError:
                            continue
                    
                except Exception as e:
                    print(f"  âŒ {filename} ë¡œë“œ ì‹¤íŒ¨: {e}")
    
    def analyze_total_transaction_volume(self):
        """ì´ê±°ë˜ëŸ‰ ë¶„ì„"""
        print("\n=== 2ë‹¨ê³„: ì´ê±°ë˜ëŸ‰ ë¶„ì„ ===")
        
        transaction_summary = []
        
        for region in self.regions:
            if not self.raw_data[region]:
                continue
                
            print(f"\nğŸ  [{region}] ê±°ë˜ëŸ‰ ë¶„ì„:")
            
            region_total = 0
            file_breakdown = []
            
            for i, df in enumerate(self.raw_data[region], 1):
                file_count = len(df)
                region_total += file_count
                file_breakdown.append(f"íŒŒì¼{i}: {file_count:,}ê±´")
                
            print(f"  ğŸ“ íŒŒì¼ë³„ ê±°ë˜ëŸ‰: {', '.join(file_breakdown)}")
            print(f"  ğŸ“ˆ ì´ ê±°ë˜ëŸ‰: {region_total:,}ê±´")
            
            # ì „ì²´ ë°ì´í„° ê²°í•©
            combined_df = pd.concat(self.raw_data[region], ignore_index=True)
            
            # ë‚ ì§œ ì»¬ëŸ¼ ì°¾ê¸°
            date_cols = [col for col in combined_df.columns if any(keyword in col.lower() 
                        for keyword in ['ë‚ ì§œ', 'date', 'ê³„ì•½', 'ê±°ë˜', 'ë…„', 'ì›”'])]
            
            print(f"  ğŸ“… ë°œê²¬ëœ ë‚ ì§œ ê´€ë ¨ ì»¬ëŸ¼: {date_cols}")
            
            # ë‚ ì§œë³„ ë¶„ì„ ì‹œë„
            if date_cols:
                self.analyze_temporal_patterns(combined_df, region, date_cols)
            
            transaction_summary.append({
                'ì§€ì—­': region,
                'ì´ê±°ë˜ëŸ‰': region_total,
                'íŒŒì¼ìˆ˜': len(self.raw_data[region]),
                'í‰ê· íŒŒì¼í¬ê¸°': region_total // len(self.raw_data[region]) if self.raw_data[region] else 0
            })
        
        # ê±°ë˜ëŸ‰ ë¹„êµ ì‹œê°í™”
        self.visualize_transaction_volume(transaction_summary)
        
        return transaction_summary
    
    def analyze_temporal_patterns(self, df, region, date_cols):
        """ì‹œê°„ë³„ íŒ¨í„´ ë¶„ì„"""
        print(f"  ğŸ• ì‹œê°„ë³„ íŒ¨í„´ ë¶„ì„:")
        
        for date_col in date_cols[:2]:  # ìƒìœ„ 2ê°œ ë‚ ì§œ ì»¬ëŸ¼ë§Œ ì‹œë„
            try:
                # ë‹¤ì–‘í•œ ë‚ ì§œ í˜•ì‹ ì‹œë„
                date_formats = ['%Y-%m-%d', '%Y.%m.%d', '%Y/%m/%d', '%Y-%m', '%Y.%m']
                
                for fmt in date_formats:
                    try:
                        df['parsed_date'] = pd.to_datetime(df[date_col], format=fmt)
                        break
                    except:
                        continue
                else:
                    # ìë™ íŒŒì‹± ì‹œë„
                    df['parsed_date'] = pd.to_datetime(df[date_col], errors='coerce')
                
                if 'parsed_date' in df.columns and not df['parsed_date'].isna().all():
                    # ì—°ë„ë³„ ì§‘ê³„
                    yearly = df.groupby(df['parsed_date'].dt.year).size()
                    print(f"    ğŸ“Š {date_col} ê¸°ì¤€ ì—°ë„ë³„ ê±°ë˜ëŸ‰:")
                    for year, count in yearly.items():
                        if pd.notna(year):
                            print(f"      {int(year)}ë…„: {count:,}ê±´")
                    
                    # ì›”ë³„ ì§‘ê³„ (ìµœê·¼ ì—°ë„)
                    if len(yearly) > 0:
                        latest_year = yearly.index.max()
                        if pd.notna(latest_year):
                            monthly = df[df['parsed_date'].dt.year == latest_year].groupby(
                                df['parsed_date'].dt.month).size()
                            print(f"    ğŸ“… {int(latest_year)}ë…„ ì›”ë³„ ê±°ë˜ëŸ‰:")
                            for month, count in monthly.items():
                                if pd.notna(month):
                                    print(f"      {int(month):2d}ì›”: {count:,}ê±´")
                    break
                    
            except Exception as e:
                print(f"    âš ï¸ {date_col} ë‚ ì§œ íŒŒì‹± ì‹¤íŒ¨: {e}")
    
    def visualize_transaction_volume(self, summary_data):
        """ê±°ë˜ëŸ‰ ì‹œê°í™”"""
        print(f"\n=== 3ë‹¨ê³„: ê±°ë˜ëŸ‰ ì‹œê°í™” ===")
        
        df_summary = pd.DataFrame(summary_data)
        
        # 1. Plotly ì¸í„°ë™í‹°ë¸Œ ë°”ì°¨íŠ¸
        fig = go.Figure()
        
        colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
        
        fig.add_trace(go.Bar(
            x=df_summary['ì§€ì—­'],
            y=df_summary['ì´ê±°ë˜ëŸ‰'],
            text=df_summary['ì´ê±°ë˜ëŸ‰'].apply(lambda x: f'{x:,}ê±´'),
            textposition='auto',
            marker_color=colors,
            hovertemplate='<b>%{x}</b><br>ì´ ê±°ë˜ëŸ‰: %{y:,}ê±´<br>íŒŒì¼ ìˆ˜: %{customdata}ê°œ<extra></extra>',
            customdata=df_summary['íŒŒì¼ìˆ˜']
        ))
        
        fig.update_layout(
            title='ğŸ  ì§€ì—­ë³„ ì´ ê±°ë˜ëŸ‰ ë¹„êµ',
            title_x=0.5,
            title_font_size=20,
            xaxis_title='ì§€ì—­',
            yaxis_title='ì´ ê±°ë˜ëŸ‰ (ê±´)',
            showlegend=False,
            height=500
        )
        
        fig.show()
        
        # 2. ê±°ë˜ëŸ‰ ë¹„ìœ¨ íŒŒì´ì°¨íŠ¸
        fig2 = go.Figure(data=[go.Pie(
            labels=df_summary['ì§€ì—­'],
            values=df_summary['ì´ê±°ë˜ëŸ‰'],
            hole=0.3,
            marker_colors=colors,
            textinfo='label+percent+value',
            texttemplate='<b>%{label}</b><br>%{percent}<br>%{value:,}ê±´'
        )])
        
        fig2.update_layout(
            title='ğŸ“Š ì§€ì—­ë³„ ê±°ë˜ëŸ‰ ë¹„ìœ¨',
            title_x=0.5,
            title_font_size=20,
            height=500
        )
        
        fig2.show()
        
        # 3. ìš”ì•½ í†µê³„
        total_transactions = df_summary['ì´ê±°ë˜ëŸ‰'].sum()
        print(f"\nğŸ“ˆ ê±°ë˜ëŸ‰ ìš”ì•½:")
        print(f"  ğŸ”¢ ì „ì²´ ê±°ë˜ëŸ‰: {total_transactions:,}ê±´")
        
        for _, row in df_summary.iterrows():
            percentage = (row['ì´ê±°ë˜ëŸ‰'] / total_transactions) * 100
            print(f"  ğŸ“ {row['ì§€ì—­']}: {row['ì´ê±°ë˜ëŸ‰']:,}ê±´ ({percentage:.1f}%)")
    
    def run_step1_analysis(self):
        """1ë‹¨ê³„ ë¶„ì„ ì‹¤í–‰"""
        print("ğŸš€ ë¶€ë™ì‚° ë°ì´í„° ë‹¨ê³„ë³„ ë¶„ì„ ì‹œì‘!")
        print("=" * 60)
        
        # ë‹¨ê³„ë³„ ì‹¤í–‰
        self.load_and_explore_raw_data()
        summary = self.analyze_total_transaction_volume()
        
        
        return summary

# ë¶„ì„ ì‹¤í–‰
analyzer = StepByStepAnalyzer()
results = analyzer.run_step1_analysis()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

class ChronologicalAnalyzer:
    def __init__(self, data_path='data/'):
        self.data_path = data_path
        self.regions = ['ê´‘ì£¼', 'ë¶€ì‚°', 'ì œì£¼']
        self.chronological_data = {}
        
    def extract_actual_periods(self):
        """ì‹¤ì œ ë°ì´í„° ê¸°ê°„ ì¶”ì¶œ ë° ì‹œê°„ìˆœ ì •ë ¬"""
        print("=== ì‹¤ì œ ë°ì´í„° ê¸°ê°„ ê¸°ì¤€ ë¶„ì„ ===")
        
        for region in self.regions:
            print(f"\nğŸ“… [{region}] ì‹¤ì œ ê¸°ê°„ë³„ ë°ì´í„° ì¶”ì¶œ:")
            
            period_data = []
            
            for i in range(1, 6):
                try:
                    filename = f"{region}_ì•„íŒŒíŠ¸_{i}.csv"
                    filepath = f"{self.data_path}{filename}"
                    
                    # íŒŒì¼ ë¡œë“œ
                    for encoding in ['utf-8', 'cp949', 'euc-kr']:
                        try:
                            df = pd.read_csv(filepath, encoding=encoding)
                            
                            # ë‚ ì§œ ë° ê°€ê²© ì²˜ë¦¬
                            processed_info = self.process_file_data(df, filename)
                            
                            if processed_info:
                                period_data.append({
                                    'original_file': filename,
                                    'file_num': i,
                                    'start_date': processed_info['start_date'],
                                    'end_date': processed_info['end_date'],
                                    'period_label': processed_info['period_label'],
                                    'valid_count': processed_info['valid_count'],
                                    'mean_price': processed_info['mean_price'],
                                    'median_price': processed_info['median_price'],
                                    'start_year_month': processed_info['start_year_month'],
                                    'end_year_month': processed_info['end_year_month'],
                                    'data': processed_info['valid_data']
                                })
                                
                                print(f"  ğŸ“„ {filename}")
                                print(f"    ğŸ“… ì‹¤ì œê¸°ê°„: {processed_info['period_label']}")
                                print(f"    ğŸ“Š ìœ íš¨ë°ì´í„°: {processed_info['valid_count']:,}ê±´")
                                print(f"    ğŸ’° í‰ê· ê°€ê²©: {processed_info['mean_price']:,.0f}ë§Œì›")
                            
                            break
                        except UnicodeDecodeError:
                            continue
                            
                except Exception as e:
                    print(f"  âŒ {filename}: {e}")
            
            # ì‹œê°„ìˆœìœ¼ë¡œ ì •ë ¬ (ì‹œì‘ ë‚ ì§œ ê¸°ì¤€)
            if period_data:
                period_data_sorted = sorted(period_data, key=lambda x: x['start_date'])
                
                print(f"\n  ğŸ• ì‹œê°„ìˆœ ì •ë ¬ ê²°ê³¼:")
                for idx, period in enumerate(period_data_sorted, 1):
                    print(f"    {idx}ìˆœìœ„: {period['period_label']} (ì›íŒŒì¼: {period['original_file']})")
                
                self.chronological_data[region] = period_data_sorted
    
    def process_file_data(self, df, filename):
        """íŒŒì¼ ë°ì´í„° ì²˜ë¦¬"""
        # ë‚ ì§œ ì»¬ëŸ¼ ì°¾ê¸° ë° íŒŒì‹±
        date_cols = [col for col in df.columns if any(keyword in col.lower() 
                    for keyword in ['ë‚ ì§œ', 'date', 'ê³„ì•½', 'ê±°ë˜', 'ë…„ì›”'])]
        
        parsed_dates = None
        for date_col in date_cols:
            try:
                # ì—¬ëŸ¬ ë‚ ì§œ í˜•ì‹ ì‹œë„
                formats = ['%Y-%m-%d', '%Y.%m.%d', '%Y/%m/%d', '%Y-%m', '%Y.%m', '%Y%m']
                
                for fmt in formats:
                    try:
                        temp_dates = pd.to_datetime(df[date_col], format=fmt, errors='coerce')
                        if temp_dates.notna().sum() / len(temp_dates) > 0.8:  # 80% ì´ìƒ ì„±ê³µ
                            parsed_dates = temp_dates
                            break
                    except:
                        continue
                
                if parsed_dates is None:
                    parsed_dates = pd.to_datetime(df[date_col], errors='coerce')
                
                if parsed_dates.notna().sum() > len(df) * 0.5:  # 50% ì´ìƒ ì„±ê³µ
                    break
                    
            except:
                continue
        
        if parsed_dates is None or parsed_dates.notna().sum() == 0:
            return None
        
        # ê°€ê²© ì»¬ëŸ¼ ì²˜ë¦¬
        price_keywords = ['ê°€ê²©', 'ê¸ˆì•¡', 'price', 'ê±°ë˜ê¸ˆì•¡', 'ë§¤ë§¤ê°€']
        price_cols = [col for col in df.columns if any(keyword in col.lower() for keyword in price_keywords)]
        
        if not price_cols:
            return None
        
        main_price_col = price_cols[0]
        
        try:
            if df[main_price_col].dtype == 'object':
                price_clean = df[main_price_col].astype(str)
                price_clean = price_clean.str.replace(',', '').str.replace('ë§Œì›', '').str.replace('ì›', '')
                price_clean = price_clean.str.extract('(\d+\.?\d*)')[0]
                cleaned_prices = pd.to_numeric(price_clean, errors='coerce')
            else:
                cleaned_prices = pd.to_numeric(df[main_price_col], errors='coerce')
        except:
            return None
        
        # ìœ íš¨í•œ ë°ì´í„°ë§Œ í•„í„°ë§
        df['ê±°ë˜ë‚ ì§œ'] = parsed_dates
        df['ê±°ë˜ê°€ê²©'] = cleaned_prices
        
        valid_data = df[(df['ê±°ë˜ë‚ ì§œ'].notna()) & 
                       (df['ê±°ë˜ê°€ê²©'].notna()) & 
                       (df['ê±°ë˜ê°€ê²©'] > 0) & 
                       (df['ê±°ë˜ê°€ê²©'] < 1000000)]  # í˜„ì‹¤ì  ë²”ìœ„
        
        if len(valid_data) == 0:
            return None
        
        # ê¸°ê°„ ì •ë³´ ê³„ì‚°
        start_date = valid_data['ê±°ë˜ë‚ ì§œ'].min()
        end_date = valid_data['ê±°ë˜ë‚ ì§œ'].max()
        
        start_ym = start_date.strftime('%Y.%m')
        end_ym = end_date.strftime('%Y.%m')
        period_label = f"{start_ym}~{end_ym}"
        
        # ê°€ê²© í†µê³„
        mean_price = valid_data['ê±°ë˜ê°€ê²©'].mean()
        median_price = valid_data['ê±°ë˜ê°€ê²©'].median()
        
        return {
            'start_date': start_date,
            'end_date': end_date,
            'start_year_month': start_ym,
            'end_year_month': end_ym,
            'period_label': period_label,
            'valid_count': len(valid_data),
            'mean_price': mean_price,
            'median_price': median_price,
            'valid_data': valid_data
        }
    
    def create_period_based_analysis(self):
        """ê¸°ê°„ ê¸°ì¤€ ë¶„ì„"""
        print("\n=== ê¸°ê°„ ê¸°ì¤€ ì¢…í•© ë¶„ì„ ===")
        
        analysis_results = {}
        
        for region in self.regions:
            if region not in self.chronological_data:
                continue
                
            periods = self.chronological_data[region]
            print(f"\nğŸ“Š [{region}] ê¸°ê°„ë³„ ê°€ê²© ì¶”ì´:")
            
            # ê¸°ê°„ë³„ ìš”ì•½
            period_summary = []
            for i, period in enumerate(periods):
                period_summary.append({
                    'chronological_order': i + 1,
                    'period_label': period['period_label'],
                    'start_date': period['start_date'],
                    'end_date': period['end_date'],
                    'duration_days': (period['end_date'] - period['start_date']).days,
                    'valid_count': period['valid_count'],
                    'mean_price': period['mean_price'],
                    'median_price': period['median_price'],
                    'original_file': period['original_file']
                })
                
                print(f"  {i+1}ê¸°: {period['period_label']} ({period['valid_count']:,}ê±´, {period['mean_price']:,.0f}ë§Œì›)")
            
            # ì¦ê°ë¥  ê³„ì‚°
            for i in range(1, len(period_summary)):
                prev_price = period_summary[i-1]['mean_price']
                curr_price = period_summary[i]['mean_price']
                change_rate = ((curr_price - prev_price) / prev_price) * 100
                period_summary[i]['price_change_rate'] = change_rate
                
                direction = "ğŸ“ˆ" if change_rate > 0 else "ğŸ“‰" if change_rate < 0 else "â¡ï¸"
                print(f"    â”” ì „ê¸° ëŒ€ë¹„: {change_rate:+.1f}% {direction}")
            
            period_summary[0]['price_change_rate'] = None  # ì²« ë²ˆì§¸ëŠ” ê¸°ì¤€
            
            analysis_results[region] = period_summary
        
        return analysis_results
    
    def create_chronological_visualization(self, analysis_results):
        """ì‹œê°„ìˆœ ì‹œê°í™”"""
        print("\n=== ì‹œê°„ìˆœ ì‹œê°í™” ìƒì„± ===")
        
        # 1. ê¸°ê°„ë³„ ê°€ê²© ì¶”ì´ (ì‹œê°„ìˆœ)
        fig1 = go.Figure()
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for region in self.regions:
            if region in analysis_results:
                data = analysis_results[region]
                
                x_values = [f"{i+1}ê¸°" for i in range(len(data))]
                y_values = [item['mean_price'] for item in data]
                period_labels = [item['period_label'] for item in data]
                counts = [item['valid_count'] for item in data]
                
                fig1.add_trace(go.Scatter(
                    x=x_values,
                    y=y_values,
                    mode='lines+markers+text',
                    name=region,
                    line=dict(color=colors[region], width=3),
                    marker=dict(size=12),
                    text=[f'{price:,.0f}ë§Œì›' for price in y_values],
                    textposition='top center',
                    textfont=dict(size=11, color=colors[region]),
                    hovertemplate=f'<b>{region}</b><br>' +
                                'ì‹¤ì œê¸°ê°„: %{customdata}<br>' +
                                'í‰ê· ê°€ê²©: %{y:,.0f}ë§Œì›<br>' +
                                'ê±°ë˜ê±´ìˆ˜: %{meta:,}ê±´<extra></extra>',
                    customdata=period_labels,
                    meta=counts
                ))
        
        fig1.update_layout(
            title='ğŸ“ˆ ì‹¤ì œ ê¸°ê°„ ê¸°ì¤€ í‰ê·  ê±°ë˜ê°€ê²© ì¶”ì´ (ì‹œê°„ìˆœ)',
            title_x=0.5,
            title_font_size=20,
            xaxis_title='ì‹œê°„ìˆœ ê¸°ê°„',
            yaxis_title='í‰ê·  ê±°ë˜ê°€ê²© (ë§Œì›)',
            height=600,
            margin=dict(t=100, b=100)
        )
        
        fig1.show()
        
        # 2. ê°€ê²© ë³€ë™ë¥  ì°¨íŠ¸
        self.create_change_rate_chart(analysis_results)
        
        # 3. ê¸°ê°„ë³„ ìƒì„¸ ì •ë³´ í…Œì´ë¸”
        self.create_comprehensive_table(analysis_results)
    
    def create_change_rate_chart(self, analysis_results):
        """ê°€ê²© ë³€ë™ë¥  ì°¨íŠ¸"""
        fig = go.Figure()
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for region in self.regions:
            if region in analysis_results:
                data = analysis_results[region]
                
                # ë³€ë™ë¥ ì´ ìˆëŠ” ë°ì´í„°ë§Œ (2ê¸°ë¶€í„°)
                change_data = [item for item in data if item['price_change_rate'] is not None]
                
                if change_data:
                    x_values = [f"{item['chronological_order']}ê¸°" for item in change_data]
                    y_values = [item['price_change_rate'] for item in change_data]
                    period_labels = [item['period_label'] for item in change_data]
                    
                    fig.add_trace(go.Bar(
                        x=x_values,
                        y=y_values,
                        name=region,
                        marker_color=colors[region],
                        text=[f'{v:+.1f}%' for v in y_values],
                        textposition='auto',
                        hovertemplate=f'<b>{region}</b><br>' +
                                    'ê¸°ê°„: %{customdata}<br>' +
                                    'ë³€ë™ë¥ : %{y:+.1f}%<extra></extra>',
                        customdata=period_labels
                    ))
        
        fig.add_hline(y=0, line=dict(color='gray', dash='dash', width=1))
        
        fig.update_layout(
            title='ğŸ“Š ê¸°ê°„ë³„ ê°€ê²© ë³€ë™ë¥  (ì „ê¸° ëŒ€ë¹„)',
            title_x=0.5,
            title_font_size=20,
            xaxis_title='ê¸°ê°„',
            yaxis_title='ê°€ê²© ë³€ë™ë¥  (%)',
            height=500,
            barmode='group'
        )
        
        fig.show()
    
    def create_comprehensive_table(self, analysis_results):
        """ì¢…í•© ì •ë³´ í…Œì´ë¸”"""
        print("\n=== ê¸°ê°„ë³„ ì¢…í•© ì •ë³´ í…Œì´ë¸” ===")
        
        # ëª¨ë“  ì§€ì—­ ë°ì´í„°ë¥¼ í•˜ë‚˜ì˜ í…Œì´ë¸”ë¡œ
        table_data = []
        
        for region in self.regions:
            if region in analysis_results:
                for item in analysis_results[region]:
                    change_str = f"{item['price_change_rate']:+.1f}%" if item['price_change_rate'] is not None else "-"
                    
                    table_data.append([
                        region,
                        f"{item['chronological_order']}ê¸°",
                        item['period_label'],
                        f"{item['duration_days']}ì¼",
                        f"{item['valid_count']:,}ê±´",
                        f"{item['mean_price']:,.0f}ë§Œì›",
                        change_str,
                        item['original_file']
                    ])
        
        if table_data:
            fig = go.Figure(data=[go.Table(
                header=dict(
                    values=['ì§€ì—­', 'ìˆœì„œ', 'ì‹¤ì œ ê¸°ê°„', 'ê¸°ê°„', 'ê±°ë˜ê±´ìˆ˜', 'í‰ê· ê°€ê²©', 'ë³€ë™ë¥ ', 'ì›ë³¸íŒŒì¼'],
                    fill_color='lightblue',
                    align='center',
                    font=dict(size=11, color='black'),
                    height=40
                ),
                cells=dict(
                    values=list(zip(*table_data)),
                    fill_color='white',
                    align='center',
                    font=dict(size=10),
                    height=32
                )
            )])
            
            fig.update_layout(
                title='ğŸ“‹ ì§€ì—­ë³„ ì‹¤ì œ ê¸°ê°„ ê¸°ì¤€ ì¢…í•© ë¶„ì„í‘œ',
                title_x=0.5,
                title_font_size=16,
                height=500,
                margin=dict(l=20, r=20, t=80, b=20)
            )
            
            fig.show()
    
    def run_chronological_analysis(self):
        """ì‹œê°„ìˆœ ì¢…í•© ë¶„ì„ ì‹¤í–‰"""
        print("ğŸ“… ì‹¤ì œ ê¸°ê°„ ê¸°ì¤€ ì‹œê°„ìˆœ ë¶„ì„ ì‹œì‘")
        print("=" * 60)
        
        self.extract_actual_periods()
        analysis_results = self.create_period_based_analysis()
        self.create_chronological_visualization(analysis_results)
        
        print(f"\nâœ… ì‹¤ì œ ê¸°ê°„ ê¸°ì¤€ ë¶„ì„ ì™„ë£Œ")
        print(f"   íŒŒì¼ ìˆœì„œê°€ ì•„ë‹Œ ì‹¤ì œ ê±°ë˜ ì‹œê¸°ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì •í™•í•œ ë¶„ì„ì„ ìˆ˜í–‰í–ˆìŠµë‹ˆë‹¤.")
        
        return analysis_results

# ì‹¤í–‰
chronological_analyzer = ChronologicalAnalyzer()
chronological_results = chronological_analyzer.run_chronological_analysis()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

class VolumePriceCorrelationAnalyzer:
    def __init__(self, data_path='data/'):
        self.data_path = data_path
        self.regions = ['ê´‘ì£¼', 'ë¶€ì‚°', 'ì œì£¼']
        self.volume_price_data = {}
        
    def load_and_process_data(self):
        """ë°ì´í„° ë¡œë“œ ë° ë…„ë„ë³„ ê±°ë˜ëŸ‰/ê°€ê²© ë¶„ì„"""
        print("=== ë…„ë„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ë¶„ì„ ===")
        
        for region in self.regions:
            print(f"\nğŸ“Š [{region}] ë°ì´í„° ì²˜ë¦¬:")
            
            all_data = []
            
            for i in range(1, 6):
                try:
                    filename = f"{region}_ì•„íŒŒíŠ¸_{i}.csv"
                    filepath = f"{self.data_path}{filename}"
                    
                    # íŒŒì¼ ë¡œë“œ
                    for encoding in ['utf-8', 'cp949', 'euc-kr']:
                        try:
                            df = pd.read_csv(filepath, encoding=encoding)
                            
                            # ë‚ ì§œ ë° ê°€ê²© ì²˜ë¦¬
                            processed_df = self.process_date_price_data(df, filename)
                            
                            if processed_df is not None and len(processed_df) > 0:
                                processed_df['ì›ë³¸íŒŒì¼'] = filename
                                all_data.append(processed_df)
                                print(f"  âœ… {filename}: {len(processed_df)}ê±´ ì²˜ë¦¬ì™„ë£Œ")
                            
                            break
                        except UnicodeDecodeError:
                            continue
                            
                except Exception as e:
                    print(f"  âŒ {filename}: {e}")
            
            if all_data:
                # ëª¨ë“  ë°ì´í„° í†µí•©
                combined_df = pd.concat(all_data, ignore_index=True)
                
                # ë…„ë„ë³„ ì§‘ê³„
                yearly_analysis = self.analyze_yearly_data(combined_df, region)
                self.volume_price_data[region] = yearly_analysis
                
                print(f"  ğŸ“ˆ ì´ {len(combined_df):,}ê±´ â†’ {len(yearly_analysis)}ë…„ê°„ ë¶„ì„ì™„ë£Œ")
    
    def process_date_price_data(self, df, filename):
        """ë‚ ì§œ ë° ê°€ê²© ë°ì´í„° ì²˜ë¦¬"""
        # ë‚ ì§œ ì»¬ëŸ¼ ì°¾ê¸°
        date_cols = [col for col in df.columns if any(keyword in col.lower() 
                    for keyword in ['ë‚ ì§œ', 'date', 'ê³„ì•½', 'ê±°ë˜', 'ë…„ì›”'])]
        
        parsed_dates = None
        for date_col in date_cols:
            try:
                formats = ['%Y-%m-%d', '%Y.%m.%d', '%Y/%m/%d', '%Y-%m', '%Y.%m', '%Y%m']
                
                for fmt in formats:
                    try:
                        temp_dates = pd.to_datetime(df[date_col], format=fmt, errors='coerce')
                        if temp_dates.notna().sum() / len(temp_dates) > 0.7:
                            parsed_dates = temp_dates
                            break
                    except:
                        continue
                
                if parsed_dates is None:
                    parsed_dates = pd.to_datetime(df[date_col], errors='coerce')
                
                if parsed_dates.notna().sum() > len(df) * 0.5:
                    break
            except:
                continue
        
        # ê°€ê²© ì»¬ëŸ¼ ì°¾ê¸°
        price_keywords = ['ê°€ê²©', 'ê¸ˆì•¡', 'price', 'ê±°ë˜ê¸ˆì•¡', 'ë§¤ë§¤ê°€']
        price_cols = [col for col in df.columns if any(keyword in col.lower() for keyword in price_keywords)]
        
        if not price_cols or parsed_dates is None:
            return None
        
        main_price_col = price_cols[0]
        
        try:
            if df[main_price_col].dtype == 'object':
                price_clean = df[main_price_col].astype(str)
                price_clean = price_clean.str.replace(',', '').str.replace('ë§Œì›', '').str.replace('ì›', '')
                price_clean = price_clean.str.extract('(\d+\.?\d*)')[0]
                cleaned_prices = pd.to_numeric(price_clean, errors='coerce')
            else:
                cleaned_prices = pd.to_numeric(df[main_price_col], errors='coerce')
        except:
            return None
        
        # ìœ íš¨í•œ ë°ì´í„°ë§Œ ì¶”ì¶œ
        df['ê±°ë˜ë‚ ì§œ'] = parsed_dates
        df['ê±°ë˜ê°€ê²©'] = cleaned_prices
        df['ê±°ë˜ë…„ë„'] = df['ê±°ë˜ë‚ ì§œ'].dt.year
        df['ê±°ë˜ì›”'] = df['ê±°ë˜ë‚ ì§œ'].dt.month
        
        valid_data = df[(df['ê±°ë˜ë‚ ì§œ'].notna()) & 
                       (df['ê±°ë˜ê°€ê²©'].notna()) & 
                       (df['ê±°ë˜ê°€ê²©'] > 0) & 
                       (df['ê±°ë˜ê°€ê²©'] < 1000000) &
                       (df['ê±°ë˜ë…„ë„'].notna())]
        
        return valid_data
    
    def analyze_yearly_data(self, df, region):
        """ë…„ë„ë³„ ë°ì´í„° ë¶„ì„"""
        print(f"\n  ğŸ“… [{region}] ë…„ë„ë³„ ë¶„ì„:")
        
        yearly_stats = df.groupby('ê±°ë˜ë…„ë„').agg({
            'ê±°ë˜ê°€ê²©': ['count', 'mean', 'median', 'std'],
            'ê±°ë˜ë‚ ì§œ': ['min', 'max']
        }).round(2)
        
        yearly_stats.columns = ['ê±°ë˜ëŸ‰', 'í‰ê· ê°€ê²©', 'ì¤‘ìœ„ê°€ê²©', 'ê°€ê²©í‘œì¤€í¸ì°¨', 'ê¸°ê°„ì‹œì‘', 'ê¸°ê°„ì¢…ë£Œ']
        
        # ì „ë…„ ëŒ€ë¹„ ì¦ê°ë¥  ê³„ì‚°
        yearly_stats['ê±°ë˜ëŸ‰ì¦ê°ë¥ '] = yearly_stats['ê±°ë˜ëŸ‰'].pct_change() * 100
        yearly_stats['í‰ê· ê°€ê²©ì¦ê°ë¥ '] = yearly_stats['í‰ê· ê°€ê²©'].pct_change() * 100
        yearly_stats['ì¤‘ìœ„ê°€ê²©ì¦ê°ë¥ '] = yearly_stats['ì¤‘ìœ„ê°€ê²©'].pct_change() * 100
        
        yearly_results = []
        
        for year, row in yearly_stats.iterrows():
            if pd.notna(year):
                year_int = int(year)
                
                result = {
                    'year': year_int,
                    'volume': int(row['ê±°ë˜ëŸ‰']),
                    'mean_price': row['í‰ê· ê°€ê²©'],
                    'median_price': row['ì¤‘ìœ„ê°€ê²©'],
                    'price_std': row['ê°€ê²©í‘œì¤€í¸ì°¨'],
                    'volume_change': row['ê±°ë˜ëŸ‰ì¦ê°ë¥ '] if pd.notna(row['ê±°ë˜ëŸ‰ì¦ê°ë¥ ']) else None,
                    'price_change': row['í‰ê· ê°€ê²©ì¦ê°ë¥ '] if pd.notna(row['í‰ê· ê°€ê²©ì¦ê°ë¥ ']) else None,
                    'median_change': row['ì¤‘ìœ„ê°€ê²©ì¦ê°ë¥ '] if pd.notna(row['ì¤‘ìœ„ê°€ê²©ì¦ê°ë¥ ']) else None,
                    'period_start': row['ê¸°ê°„ì‹œì‘'].strftime('%Y.%m.%d'),
                    'period_end': row['ê¸°ê°„ì¢…ë£Œ'].strftime('%Y.%m.%d')
                }
                
                yearly_results.append(result)
                
                # ì¶œë ¥
                vol_change_str = f" ({result['volume_change']:+.1f}%)" if result['volume_change'] is not None else ""
                price_change_str = f" ({result['price_change']:+.1f}%)" if result['price_change'] is not None else ""
                
                print(f"    {year_int}ë…„: {result['volume']:,}ê±´{vol_change_str}, {result['mean_price']:,.0f}ë§Œì›{price_change_str}")
        
        return yearly_results
    
    def create_dual_axis_visualization(self):
        """ì´ì¤‘ì¶• ì‹œê°í™” (ê±°ë˜ëŸ‰ + ê°€ê²©)"""
        print("\n=== ê±°ë˜ëŸ‰ vs ê°€ê²© ì´ì¤‘ì¶• ì‹œê°í™” ===")
        
        # ì§€ì—­ë³„ ì„œë¸Œí”Œë¡¯
        fig = make_subplots(
            rows=len(self.regions), cols=1,
            subplot_titles=[f'{region} - ë…„ë„ë³„ ê±°ë˜ëŸ‰ vs í‰ê· ê°€ê²©' for region in self.regions],
            specs=[[{"secondary_y": True}] for _ in self.regions],
            vertical_spacing=0.1
        )
        
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for row, region in enumerate(self.regions, 1):
            if region in self.volume_price_data:
                data = self.volume_price_data[region]
                
                years = [d['year'] for d in data]
                volumes = [d['volume'] for d in data]
                prices = [d['mean_price'] for d in data]
                
                # ê±°ë˜ëŸ‰ (ë§‰ëŒ€ê·¸ë˜í”„)
                fig.add_trace(
                    go.Bar(
                        x=years,
                        y=volumes,
                        name=f'{region} ê±°ë˜ëŸ‰',
                        marker_color=colors[region],
                        opacity=0.7,
                        text=[f'{v:,}ê±´' for v in volumes],
                        textposition='auto',
                        yaxis='y',
                        hovertemplate=f'<b>{region} ê±°ë˜ëŸ‰</b><br>ë…„ë„: %{{x}}<br>ê±°ë˜ëŸ‰: %{{y:,}}ê±´<extra></extra>',
                        showlegend=(row==1)
                    ),
                    row=row, col=1,
                    secondary_y=False
                )
                
                # í‰ê· ê°€ê²© (ì„ ê·¸ë˜í”„)
                fig.add_trace(
                    go.Scatter(
                        x=years,
                        y=prices,
                        mode='lines+markers+text',
                        name=f'{region} í‰ê· ê°€ê²©',
                        line=dict(color=colors[region], width=4),
                        marker=dict(size=10, color='white', line=dict(color=colors[region], width=2)),
                        text=[f'{p:,.0f}ë§Œì›' for p in prices],
                        textposition='top center',
                        textfont=dict(color=colors[region], size=10),
                        yaxis='y2',
                        hovertemplate=f'<b>{region} í‰ê· ê°€ê²©</b><br>ë…„ë„: %{{x}}<br>í‰ê· ê°€ê²©: %{{y:,.0f}}ë§Œì›<extra></extra>',
                        showlegend=(row==1)
                    ),
                    row=row, col=1,
                    secondary_y=True
                )
        
        # ì¶• ë¼ë²¨ ì„¤ì •
        for i in range(1, len(self.regions) + 1):
            fig.update_yaxes(title_text="ê±°ë˜ëŸ‰ (ê±´)", secondary_y=False, row=i, col=1)
            fig.update_yaxes(title_text="í‰ê· ê°€ê²© (ë§Œì›)", secondary_y=True, row=i, col=1)
        
        fig.update_layout(
            title='ğŸ“Š ë…„ë„ë³„ ê±°ë˜ëŸ‰ vs í‰ê· ê°€ê²© ë™ì‹œ ë¶„ì„',
            title_x=0.5,
            title_font_size=20,
            height=300 * len(self.regions),
            margin=dict(t=100, b=50)
        )
        
        fig.show()
    
    def create_correlation_analysis(self):
        """ìƒê´€ê´€ê³„ ë¶„ì„"""
        print("\n=== ê±°ë˜ëŸ‰-ê°€ê²© ìƒê´€ê´€ê³„ ë¶„ì„ ===")
        
        # 1. ì‚°ì ë„ (ê±°ë˜ëŸ‰ vs ê°€ê²©)
        fig1 = go.Figure()
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for region in self.regions:
            if region in self.volume_price_data:
                data = self.volume_price_data[region]
                
                volumes = [d['volume'] for d in data]
                prices = [d['mean_price'] for d in data]
                years = [d['year'] for d in data]
                
                # ìƒê´€ê³„ìˆ˜ ê³„ì‚°
                if len(volumes) > 2:
                    correlation = np.corrcoef(volumes, prices)[0, 1]
                    
                    fig1.add_trace(go.Scatter(
                        x=volumes,
                        y=prices,
                        mode='markers+text',
                        name=f'{region} (ìƒê´€ê³„ìˆ˜: {correlation:.2f})',
                        marker=dict(size=12, color=colors[region]),
                        text=[str(y) for y in years],
                        textposition='top center',
                        hovertemplate=f'<b>{region}</b><br>ë…„ë„: %{{text}}<br>ê±°ë˜ëŸ‰: %{{x:,}}ê±´<br>í‰ê· ê°€ê²©: %{{y:,.0f}}ë§Œì›<extra></extra>'
                    ))
        
        fig1.update_layout(
            title='ğŸ“ˆ ê±°ë˜ëŸ‰ vs í‰ê· ê°€ê²© ìƒê´€ê´€ê³„ (ì‚°ì ë„)',
            title_x=0.5,
            title_font_size=18,
            xaxis_title='ë…„ê°„ ê±°ë˜ëŸ‰ (ê±´)',
            yaxis_title='í‰ê· ê°€ê²© (ë§Œì›)',
            height=500
        )
        
        fig1.show()
        
        # 2. ì¦ê°ë¥  ë¹„êµ
        self.create_change_rate_comparison()
    
    def create_change_rate_comparison(self):
        """ì¦ê°ë¥  ë¹„êµ ë¶„ì„"""
        fig = make_subplots(
            rows=1, cols=2,
            subplot_titles=('ê±°ë˜ëŸ‰ ì¦ê°ë¥ ', 'í‰ê· ê°€ê²© ì¦ê°ë¥ '),
            specs=[[{"secondary_y": False}, {"secondary_y": False}]]
        )
        
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for region in self.regions:
            if region in self.volume_price_data:
                data = self.volume_price_data[region]
                
                # ì¦ê°ë¥ ì´ ìˆëŠ” ë°ì´í„°ë§Œ (2ë…„ì°¨ë¶€í„°)
                change_data = [d for d in data if d['volume_change'] is not None and d['price_change'] is not None]
                
                if change_data:
                    years = [d['year'] for d in change_data]
                    volume_changes = [d['volume_change'] for d in change_data]
                    price_changes = [d['price_change'] for d in change_data]
                    
                    # ê±°ë˜ëŸ‰ ì¦ê°ë¥ 
                    fig.add_trace(
                        go.Bar(
                            x=years,
                            y=volume_changes,
                            name=f'{region}',
                            marker_color=colors[region],
                            text=[f'{v:+.1f}%' for v in volume_changes],
                            textposition='auto',
                            showlegend=True
                        ),
                        row=1, col=1
                    )
                    
                    # ê°€ê²© ì¦ê°ë¥ 
                    fig.add_trace(
                        go.Bar(
                            x=years,
                            y=price_changes,
                            name=f'{region}',
                            marker_color=colors[region],
                            text=[f'{v:+.1f}%' for v in price_changes],
                            textposition='auto',
                            showlegend=False
                        ),
                        row=1, col=2
                    )
        
        # 0% ê¸°ì¤€ì„ 
        fig.add_hline(y=0, line=dict(color='gray', dash='dash', width=1), row=1, col=1)
        fig.add_hline(y=0, line=dict(color='gray', dash='dash', width=1), row=1, col=2)
        
        fig.update_layout(
            title='ğŸ“Š ë…„ë„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ì¦ê°ë¥  ë¹„êµ',
            title_x=0.5,
            title_font_size=18,
            height=500
        )
        
        fig.update_yaxes(title_text="ê±°ë˜ëŸ‰ ì¦ê°ë¥  (%)", row=1, col=1)
        fig.update_yaxes(title_text="ê°€ê²© ì¦ê°ë¥  (%)", row=1, col=2)
        
        fig.show()
    
    def create_summary_table(self):
        """ë…„ë„ë³„ ì¢…í•© ìš”ì•½ í…Œì´ë¸”"""
        print("\n=== ë…„ë„ë³„ ì¢…í•© ìš”ì•½ í…Œì´ë¸” ===")
        
        # ëª¨ë“  ì§€ì—­ ë°ì´í„°ë¥¼ í•˜ë‚˜ì˜ í…Œì´ë¸”ë¡œ
        table_data = []
        
        for region in self.regions:
            if region in self.volume_price_data:
                for item in self.volume_price_data[region]:
                    vol_change_str = f"{item['volume_change']:+.1f}%" if item['volume_change'] is not None else "-"
                    price_change_str = f"{item['price_change']:+.1f}%" if item['price_change'] is not None else "-"
                    
                    table_data.append([
                        region,
                        item['year'],
                        f"{item['volume']:,}ê±´",
                        vol_change_str,
                        f"{item['mean_price']:,.0f}ë§Œì›",
                        price_change_str,
                        f"{item['median_price']:,.0f}ë§Œì›"
                    ])
        
        if table_data:
            fig = go.Figure(data=[go.Table(
                header=dict(
                    values=['ì§€ì—­', 'ë…„ë„', 'ê±°ë˜ëŸ‰', 'ê±°ë˜ëŸ‰ì¦ê°ë¥ ', 'í‰ê· ê°€ê²©', 'ê°€ê²©ì¦ê°ë¥ ', 'ì¤‘ìœ„ê°€ê²©'],
                    fill_color='lightblue',
                    align='center',
                    font=dict(size=12, color='black'),
                    height=40
                ),
                cells=dict(
                    values=list(zip(*table_data)),
                    fill_color='white',
                    align='center',
                    font=dict(size=11),
                    height=32
                )
            )])
            
            fig.update_layout(
                title='ğŸ“‹ ë…„ë„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ì¢…í•© ìš”ì•½í‘œ',
                title_x=0.5,
                title_font_size=16,
                height=600,
                margin=dict(l=20, r=20, t=80, b=20)
            )
            
            fig.show()
    
    def analyze_market_insights(self):
        """ì‹œì¥ ì¸ì‚¬ì´íŠ¸ ë¶„ì„"""
        print("\n=== ğŸ” ì‹œì¥ ë¶„ì„ ì¸ì‚¬ì´íŠ¸ ===")
        
        for region in self.regions:
            if region not in self.volume_price_data:
                continue
                
            data = self.volume_price_data[region]
            print(f"\nğŸ  [{region}] ì‹œì¥ íŠ¹ì„±:")
            
            if len(data) < 2:
                print("  âš ï¸ ë¶„ì„ ë°ì´í„° ë¶€ì¡±")
                continue
            
            # ê±°ë˜ëŸ‰ê³¼ ê°€ê²©ì˜ ìƒê´€ê´€ê³„
            volumes = [d['volume'] for d in data]
            prices = [d['mean_price'] for d in data]
            
            if len(volumes) > 2:
                correlation = np.corrcoef(volumes, prices)[0, 1]
                
                if correlation > 0.5:
                    print(f"  ğŸ“ˆ ê°•í•œ ì–‘ì˜ ìƒê´€ê´€ê³„ ({correlation:.2f})")
                    print("     â†’ ê±°ë˜ëŸ‰ ì¦ê°€ ì‹œ ê°€ê²©ë„ ìƒìŠ¹í•˜ëŠ” ê²½í–¥")
                elif correlation < -0.5:
                    print(f"  ğŸ“‰ ê°•í•œ ìŒì˜ ìƒê´€ê´€ê³„ ({correlation:.2f})")
                    print("     â†’ ê±°ë˜ëŸ‰ ì¦ê°€ ì‹œ ê°€ê²©ì€ í•˜ë½í•˜ëŠ” ê²½í–¥ (ë§¤ë¬¼ ê¸‰ì¦)")
                else:
                    print(f"  ğŸ“Š ì•½í•œ ìƒê´€ê´€ê³„ ({correlation:.2f})")
                    print("     â†’ ê±°ë˜ëŸ‰ê³¼ ê°€ê²©ì´ ë…ë¦½ì ìœ¼ë¡œ ì›€ì§ì„")
            
            # ìµœê·¼ 3ë…„ íŠ¸ë Œë“œ
            recent_data = data[-3:] if len(data) >= 3 else data
            
            volume_trend = "ì¦ê°€" if recent_data[-1]['volume'] > recent_data[0]['volume'] else "ê°ì†Œ"
            price_trend = "ìƒìŠ¹" if recent_data[-1]['mean_price'] > recent_data[0]['mean_price'] else "í•˜ë½"
            
            print(f"  ğŸ“… ìµœê·¼ íŠ¸ë Œë“œ: ê±°ë˜ëŸ‰ {volume_trend}, ê°€ê²© {price_trend}")
            
            # ê°€ê²© ëŒ€ë¹„ ê±°ë˜ëŸ‰ íŒ¨í„´
            max_price_year = max(data, key=lambda x: x['mean_price'])
            max_volume_year = max(data, key=lambda x: x['volume'])
            
            print(f"  ğŸ”¥ ìµœê³ ê°€ê²©: {max_price_year['year']}ë…„ ({max_price_year['mean_price']:,.0f}ë§Œì›)")
            print(f"  ğŸ“Š ìµœë‹¤ê±°ë˜: {max_volume_year['year']}ë…„ ({max_volume_year['volume']:,}ê±´)")
            
            if max_price_year['year'] == max_volume_year['year']:
                print("     ğŸ’¡ ê°€ê²© ìµœê³ ì ê³¼ ê±°ë˜ëŸ‰ ìµœê³ ì ì´ ë™ì¼ â†’ ê³¼ì—´ ì–‘ìƒ")
            else:
                print("     ğŸ’¡ ê°€ê²© ìµœê³ ì ê³¼ ê±°ë˜ëŸ‰ ìµœê³ ì ì´ ìƒì´ â†’ ì‹œì°¨ ì¡´ì¬")
    
    def run_volume_price_analysis(self):
        """ì „ì²´ ë¶„ì„ ì‹¤í–‰"""
        print("ğŸ“Š ë…„ë„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ë™ì‹œ ë¶„ì„ ì‹œì‘")
        print("=" * 60)
        
        self.load_and_process_data()
        self.create_dual_axis_visualization()
        self.create_correlation_analysis()
        self.create_summary_table()
        self.analyze_market_insights()
        
        print(f"\nâœ… ê±°ë˜ëŸ‰ vs ê°€ê²© ë¶„ì„ ì™„ë£Œ")
        print(f"   ë…„ë„ë³„ ê±°ë˜ëŸ‰ê³¼ ê°€ê²©ì˜ ìƒê´€ê´€ê³„ë¥¼ ì¢…í•©ì ìœ¼ë¡œ ë¶„ì„í–ˆìŠµë‹ˆë‹¤.")
        
        return self.volume_price_data

# ë¶„ì„ ì‹¤í–‰
volume_price_analyzer = VolumePriceCorrelationAnalyzer()
correlation_results = volume_price_analyzer.run_volume_price_analysis()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

class PeriodBasedVolumePriceAnalyzer:
    def __init__(self, data_path='data/'):
        self.data_path = data_path
        self.regions = ['ê´‘ì£¼', 'ë¶€ì‚°', 'ì œì£¼']
        self.period_data = {}
        
    def extract_actual_periods(self):
        """ì‹¤ì œ ë°ì´í„° ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ ë° ê°€ê²© ì¶”ì¶œ"""
        print("=== ì‹¤ì œ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ë¶„ì„ ===")
        
        for region in self.regions:
            print(f"\nğŸ“Š [{region}] ì‹¤ì œ ê¸°ê°„ë³„ ë°ì´í„° ì¶”ì¶œ:")
            
            period_analysis = []
            
            for i in range(1, 6):
                try:
                    filename = f"{region}_ì•„íŒŒíŠ¸_{i}.csv"
                    filepath = f"{self.data_path}{filename}"
                    
                    # íŒŒì¼ ë¡œë“œ
                    for encoding in ['utf-8', 'cp949', 'euc-kr']:
                        try:
                            df = pd.read_csv(filepath, encoding=encoding)
                            
                            # ì‹¤ì œ ê¸°ê°„ ë° ê±°ë˜ëŸ‰/ê°€ê²© ë¶„ì„
                            period_info = self.analyze_file_period(df, filename, i)
                            
                            if period_info:
                                period_analysis.append(period_info)
                                
                                print(f"  ğŸ“„ {filename}")
                                print(f"    ğŸ“… ì‹¤ì œê¸°ê°„: {period_info['period_label']}")
                                print(f"    ğŸ“Š ê±°ë˜ëŸ‰: {period_info['volume']:,}ê±´")
                                print(f"    ğŸ’° í‰ê· ê°€ê²©: {period_info['mean_price']:,.0f}ë§Œì›")
                                print(f"    ğŸ’° ì¤‘ìœ„ê°€ê²©: {period_info['median_price']:,.0f}ë§Œì›")
                            
                            break
                        except UnicodeDecodeError:
                            continue
                            
                except Exception as e:
                    print(f"  âŒ {filename}: {e}")
            
            # ì‹¤ì œ ê¸°ê°„ ì‹œì‘ì¼ ê¸°ì¤€ìœ¼ë¡œ ì‹œê°„ìˆœ ì •ë ¬
            if period_analysis:
                period_analysis_sorted = sorted(period_analysis, key=lambda x: x['start_date'])
                
                print(f"\n  ğŸ• ì‹œê°„ìˆœ ì •ë ¬ ê²°ê³¼:")
                for idx, period in enumerate(period_analysis_sorted, 1):
                    print(f"    {idx}ìˆœìœ„: {period['period_label']} (ì›íŒŒì¼: {period['filename']})")
                
                # ì „ê¸° ëŒ€ë¹„ ì¦ê°ë¥  ê³„ì‚°
                for i in range(1, len(period_analysis_sorted)):
                    prev_period = period_analysis_sorted[i-1]
                    curr_period = period_analysis_sorted[i]
                    
                    # ê±°ë˜ëŸ‰ ì¦ê°ë¥ 
                    volume_change = ((curr_period['volume'] - prev_period['volume']) / prev_period['volume']) * 100
                    curr_period['volume_change'] = volume_change
                    
                    # ê°€ê²© ì¦ê°ë¥ 
                    price_change = ((curr_period['mean_price'] - prev_period['mean_price']) / prev_period['mean_price']) * 100
                    curr_period['price_change'] = price_change
                    
                    median_change = ((curr_period['median_price'] - prev_period['median_price']) / prev_period['median_price']) * 100
                    curr_period['median_change'] = median_change
                
                # ì²« ë²ˆì§¸ ê¸°ê°„ì€ ì¦ê°ë¥  ì—†ìŒ
                period_analysis_sorted[0]['volume_change'] = None
                period_analysis_sorted[0]['price_change'] = None
                period_analysis_sorted[0]['median_change'] = None
                
                self.period_data[region] = period_analysis_sorted
    
    def analyze_file_period(self, df, filename, file_num):
        """ê°œë³„ íŒŒì¼ì˜ ì‹¤ì œ ê¸°ê°„ ë° ê±°ë˜ëŸ‰/ê°€ê²© ë¶„ì„"""
        # ë‚ ì§œ ì»¬ëŸ¼ ì°¾ê¸° ë° íŒŒì‹±
        date_cols = [col for col in df.columns if any(keyword in col.lower() 
                    for keyword in ['ë‚ ì§œ', 'date', 'ê³„ì•½', 'ê±°ë˜', 'ë…„ì›”'])]
        
        parsed_dates = None
        for date_col in date_cols:
            try:
                formats = ['%Y-%m-%d', '%Y.%m.%d', '%Y/%m/%d', '%Y-%m', '%Y.%m', '%Y%m']
                
                for fmt in formats:
                    try:
                        temp_dates = pd.to_datetime(df[date_col], format=fmt, errors='coerce')
                        if temp_dates.notna().sum() / len(temp_dates) > 0.7:  # 70% ì´ìƒ ì„±ê³µ
                            parsed_dates = temp_dates
                            break
                    except:
                        continue
                
                if parsed_dates is None:
                    parsed_dates = pd.to_datetime(df[date_col], errors='coerce')
                
                if parsed_dates.notna().sum() > len(df) * 0.5:  # 50% ì´ìƒ ì„±ê³µ
                    break
            except:
                continue
        
        # ê°€ê²© ì»¬ëŸ¼ ì°¾ê¸° ë° íŒŒì‹±
        price_keywords = ['ê°€ê²©', 'ê¸ˆì•¡', 'price', 'ê±°ë˜ê¸ˆì•¡', 'ë§¤ë§¤ê°€']
        price_cols = [col for col in df.columns if any(keyword in col.lower() for keyword in price_keywords)]
        
        if not price_cols or parsed_dates is None:
            return None
        
        main_price_col = price_cols[0]
        
        try:
            if df[main_price_col].dtype == 'object':
                price_clean = df[main_price_col].astype(str)
                price_clean = price_clean.str.replace(',', '').str.replace('ë§Œì›', '').str.replace('ì›', '')
                price_clean = price_clean.str.extract('(\d+\.?\d*)')[0]
                cleaned_prices = pd.to_numeric(price_clean, errors='coerce')
            else:
                cleaned_prices = pd.to_numeric(df[main_price_col], errors='coerce')
        except:
            return None
        
        # ìœ íš¨í•œ ë°ì´í„°ë§Œ í•„í„°ë§
        df['ê±°ë˜ë‚ ì§œ'] = parsed_dates
        df['ê±°ë˜ê°€ê²©'] = cleaned_prices
        
        valid_data = df[(df['ê±°ë˜ë‚ ì§œ'].notna()) & 
                       (df['ê±°ë˜ê°€ê²©'].notna()) & 
                       (df['ê±°ë˜ê°€ê²©'] > 0) & 
                       (df['ê±°ë˜ê°€ê²©'] < 1000000)]
        
        if len(valid_data) < 50:  # ìµœì†Œ 50ê±´ ì´ìƒ
            return None
        
        # ì‹¤ì œ ê¸°ê°„ ê³„ì‚°
        start_date = valid_data['ê±°ë˜ë‚ ì§œ'].min()
        end_date = valid_data['ê±°ë˜ë‚ ì§œ'].max()
        
        # ê¸°ê°„ ë¼ë²¨ ìƒì„± (ë…„.ì›” í˜•íƒœ)
        start_ym = f"{start_date.year}.{start_date.month}"
        end_ym = f"{end_date.year}.{end_date.month}"
        period_label = f"{start_ym}~{end_ym}"
        
        # ê±°ë˜ëŸ‰ ë° ê°€ê²© í†µê³„
        volume = len(valid_data)
        mean_price = valid_data['ê±°ë˜ê°€ê²©'].mean()
        median_price = valid_data['ê±°ë˜ê°€ê²©'].median()
        price_std = valid_data['ê±°ë˜ê°€ê²©'].std()
        
        return {
            'filename': filename,
            'file_num': file_num,
            'start_date': start_date,
            'end_date': end_date,
            'period_label': period_label,
            'start_ym': start_ym,
            'end_ym': end_ym,
            'volume': volume,
            'mean_price': mean_price,
            'median_price': median_price,
            'price_std': price_std,
            'duration_days': (end_date - start_date).days
        }
    
    def create_period_dual_axis_visualization(self):
        """ì‹¤ì œ ê¸°ê°„ë³„ ì´ì¤‘ì¶• ì‹œê°í™”"""
        print("\n=== ì‹¤ì œ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ì´ì¤‘ì¶• ì‹œê°í™” ===")
        
        # ì§€ì—­ë³„ ì„œë¸Œí”Œë¡¯
        fig = make_subplots(
            rows=len(self.regions), cols=1,
            subplot_titles=[f'{region} - ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs í‰ê· ê°€ê²© (ì‹¤ì œ ê¸°ê°„ ê¸°ì¤€)' for region in self.regions],
            specs=[[{"secondary_y": True}] for _ in self.regions],
            vertical_spacing=0.15
        )
        
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for row, region in enumerate(self.regions, 1):
            if region in self.period_data:
                periods = self.period_data[region]
                
                x_labels = [f"{i+1}ê¸°\n({p['period_label']})" for i, p in enumerate(periods)]
                volumes = [p['volume'] for p in periods]
                prices = [p['mean_price'] for p in periods]
                period_labels = [p['period_label'] for p in periods]
                
                # ê±°ë˜ëŸ‰ (ë§‰ëŒ€ê·¸ë˜í”„)
                fig.add_trace(
                    go.Bar(
                        x=list(range(len(periods))),
                        y=volumes,
                        name=f'{region} ê±°ë˜ëŸ‰',
                        marker_color=colors[region],
                        opacity=0.7,
                        text=[f'{v:,}ê±´' for v in volumes],
                        textposition='auto',
                        hovertemplate=f'<b>{region} ê±°ë˜ëŸ‰</b><br>ê¸°ê°„: %{{customdata}}<br>ê±°ë˜ëŸ‰: %{{y:,}}ê±´<extra></extra>',
                        customdata=period_labels,
                        showlegend=(row==1)
                    ),
                    row=row, col=1,
                    secondary_y=False
                )
                
                # í‰ê· ê°€ê²© (ì„ ê·¸ë˜í”„)
                fig.add_trace(
                    go.Scatter(
                        x=list(range(len(periods))),
                        y=prices,
                        mode='lines+markers+text',
                        name=f'{region} í‰ê· ê°€ê²©',
                        line=dict(color=colors[region], width=4),
                        marker=dict(size=12, color='white', line=dict(color=colors[region], width=3)),
                        text=[f'{p:,.0f}ë§Œì›' for p in prices],
                        textposition='top center',
                        textfont=dict(color=colors[region], size=11, family="Arial Black"),
                        hovertemplate=f'<b>{region} í‰ê· ê°€ê²©</b><br>ê¸°ê°„: %{{customdata}}<br>í‰ê· ê°€ê²©: %{{y:,.0f}}ë§Œì›<extra></extra>',
                        customdata=period_labels,
                        showlegend=(row==1)
                    ),
                    row=row, col=1,
                    secondary_y=True
                )
                
                # xì¶• ë¼ë²¨ ì„¤ì • (ê¸°ê°„ í‘œì‹œ)
                fig.update_xaxes(
                    tickmode='array',
                    tickvals=list(range(len(periods))),
                    ticktext=x_labels,
                    row=row, col=1
                )
        
        # ì¶• ë¼ë²¨ ì„¤ì •
        for i in range(1, len(self.regions) + 1):
            fig.update_yaxes(title_text="ê±°ë˜ëŸ‰ (ê±´)", secondary_y=False, row=i, col=1)
            fig.update_yaxes(title_text="í‰ê· ê°€ê²© (ë§Œì›)", secondary_y=True, row=i, col=1)
        
        fig.update_layout(
            title='ğŸ“Š ì‹¤ì œ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs í‰ê· ê°€ê²© ë™ì‹œ ë¶„ì„',
            title_x=0.5,
            title_font_size=20,
            height=350 * len(self.regions),
            margin=dict(t=100, b=80)
        )
        
        fig.show()
    
    def create_period_correlation_analysis(self):
        """ì‹¤ì œ ê¸°ê°„ë³„ ìƒê´€ê´€ê³„ ë¶„ì„"""
        print("\n=== ê¸°ê°„ë³„ ê±°ë˜ëŸ‰-ê°€ê²© ìƒê´€ê´€ê³„ ë¶„ì„ ===")
        
        # ì‚°ì ë„
        fig1 = go.Figure()
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for region in self.regions:
            if region in self.period_data:
                periods = self.period_data[region]
                
                volumes = [p['volume'] for p in periods]
                prices = [p['mean_price'] for p in periods]
                period_labels = [p['period_label'] for p in periods]
                
                # ìƒê´€ê³„ìˆ˜ ê³„ì‚°
                if len(volumes) > 2:
                    correlation = np.corrcoef(volumes, prices)[0, 1]
                    
                    fig1.add_trace(go.Scatter(
                        x=volumes,
                        y=prices,
                        mode='markers+text',
                        name=f'{region} (ìƒê´€ê³„ìˆ˜: {correlation:.2f})',
                        marker=dict(size=15, color=colors[region]),
                        text=period_labels,
                        textposition='top center',
                        textfont=dict(size=10),
                        hovertemplate=f'<b>{region}</b><br>ê¸°ê°„: %{{text}}<br>ê±°ë˜ëŸ‰: %{{x:,}}ê±´<br>í‰ê· ê°€ê²©: %{{y:,.0f}}ë§Œì›<extra></extra>'
                    ))
        
        fig1.update_layout(
            title='ğŸ“ˆ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs í‰ê· ê°€ê²© ìƒê´€ê´€ê³„ (ì‚°ì ë„)',
            title_x=0.5,
            title_font_size=18,
            xaxis_title='ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ (ê±´)',
            yaxis_title='í‰ê· ê°€ê²© (ë§Œì›)',
            height=500
        )
        
        fig1.show()
        
        # ì¦ê°ë¥  ë¹„êµ
        self.create_period_change_rate_comparison()
    
    def create_period_change_rate_comparison(self):
        """ê¸°ê°„ë³„ ì¦ê°ë¥  ë¹„êµ"""
        fig = make_subplots(
            rows=1, cols=2,
            subplot_titles=('ê±°ë˜ëŸ‰ ì¦ê°ë¥  (ì „ê¸° ëŒ€ë¹„)', 'í‰ê· ê°€ê²© ì¦ê°ë¥  (ì „ê¸° ëŒ€ë¹„)'),
            specs=[[{"secondary_y": False}, {"secondary_y": False}]]
        )
        
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for region in self.regions:
            if region in self.period_data:
                periods = self.period_data[region]
                
                # ì¦ê°ë¥ ì´ ìˆëŠ” ë°ì´í„°ë§Œ (2ê¸°ë¶€í„°)
                change_data = [p for p in periods if p['volume_change'] is not None and p['price_change'] is not None]
                
                if change_data:
                    x_labels = [f"{periods.index(p)+1}ê¸°\n({p['period_label']})" for p in change_data]
                    volume_changes = [p['volume_change'] for p in change_data]
                    price_changes = [p['price_change'] for p in change_data]
                    
                    # ê±°ë˜ëŸ‰ ì¦ê°ë¥ 
                    fig.add_trace(
                        go.Bar(
                            x=x_labels,
                            y=volume_changes,
                            name=f'{region}',
                            marker_color=colors[region],
                            text=[f'{v:+.1f}%' for v in volume_changes],
                            textposition='auto',
                            showlegend=True
                        ),
                        row=1, col=1
                    )
                    
                    # ê°€ê²© ì¦ê°ë¥ 
                    fig.add_trace(
                        go.Bar(
                            x=x_labels,
                            y=price_changes,
                            name=f'{region}',
                            marker_color=colors[region],
                            text=[f'{v:+.1f}%' for v in price_changes],
                            textposition='auto',
                            showlegend=False
                        ),
                        row=1, col=2
                    )
        
        # 0% ê¸°ì¤€ì„ 
        fig.add_hline(y=0, line=dict(color='gray', dash='dash', width=1), row=1, col=1)
        fig.add_hline(y=0, line=dict(color='gray', dash='dash', width=1), row=1, col=2)
        
        fig.update_layout(
            title='ğŸ“Š ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ì¦ê°ë¥  ë¹„êµ (ì „ê¸° ëŒ€ë¹„)',
            title_x=0.5,
            title_font_size=18,
            height=500
        )
        
        fig.update_yaxes(title_text="ê±°ë˜ëŸ‰ ì¦ê°ë¥  (%)", row=1, col=1)
        fig.update_yaxes(title_text="ê°€ê²© ì¦ê°ë¥  (%)", row=1, col=2)
        
        fig.show()
    
    def create_comprehensive_period_table(self):
        """ì‹¤ì œ ê¸°ê°„ë³„ ì¢…í•© í…Œì´ë¸”"""
        print("\n=== ì‹¤ì œ ê¸°ê°„ë³„ ì¢…í•© ìš”ì•½ í…Œì´ë¸” ===")
        
        table_data = []
        
        for region in self.regions:
            if region in self.period_data:
                for i, period in enumerate(self.period_data[region]):
                    vol_change_str = f"{period['volume_change']:+.1f}%" if period['volume_change'] is not None else "-"
                    price_change_str = f"{period['price_change']:+.1f}%" if period['price_change'] is not None else "-"
                    median_change_str = f"{period['median_change']:+.1f}%" if period['median_change'] is not None else "-"
                    
                    table_data.append([
                        region,
                        f"{i+1}ê¸°",
                        period['period_label'],
                        f"{period['duration_days']}ì¼",
                        f"{period['volume']:,}ê±´",
                        vol_change_str,
                        f"{period['mean_price']:,.0f}ë§Œì›",
                        price_change_str,
                        f"{period['median_price']:,.0f}ë§Œì›",
                        period['filename']
                    ])
        
        if table_data:
            fig = go.Figure(data=[go.Table(
                header=dict(
                    values=['ì§€ì—­', 'ìˆœì„œ', 'ì‹¤ì œ ê¸°ê°„', 'ê¸°ê°„', 'ê±°ë˜ëŸ‰', 'ê±°ë˜ëŸ‰ì¦ê°ë¥ ', 'í‰ê· ê°€ê²©', 'ê°€ê²©ì¦ê°ë¥ ', 'ì¤‘ìœ„ê°€ê²©', 'ì›ë³¸íŒŒì¼'],
                    fill_color='lightblue',
                    align='center',
                    font=dict(size=11, color='black'),
                    height=45
                ),
                cells=dict(
                    values=list(zip(*table_data)),
                    fill_color='white',
                    align='center',
                    font=dict(size=10),
                    height=35
                )
            )])
            
            fig.update_layout(
                title='ğŸ“‹ ì‹¤ì œ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ì¢…í•© ë¶„ì„í‘œ',
                title_x=0.5,
                title_font_size=16,
                height=600,
                margin=dict(l=20, r=20, t=80, b=20)
            )
            
            fig.show()
    
    def analyze_period_market_insights(self):
        """ì‹¤ì œ ê¸°ê°„ë³„ ì‹œì¥ ì¸ì‚¬ì´íŠ¸"""
        print("\n=== ğŸ” ì‹¤ì œ ê¸°ê°„ë³„ ì‹œì¥ ë¶„ì„ ì¸ì‚¬ì´íŠ¸ ===")
        
        for region in self.regions:
            if region not in self.period_data:
                continue
                
            periods = self.period_data[region]
            print(f"\nğŸ  [{region}] ê¸°ê°„ë³„ ì‹œì¥ íŠ¹ì„±:")
            
            if len(periods) < 2:
                print("  âš ï¸ ë¶„ì„ ê¸°ê°„ ë¶€ì¡±")
                continue
            
            # ê±°ë˜ëŸ‰ê³¼ ê°€ê²©ì˜ ìƒê´€ê´€ê³„
            volumes = [p['volume'] for p in periods]
            prices = [p['mean_price'] for p in periods]
            
            if len(volumes) > 2:
                correlation = np.corrcoef(volumes, prices)[0, 1]
                
                if correlation > 0.5:
                    print(f"  ğŸ“ˆ ê°•í•œ ì–‘ì˜ ìƒê´€ê´€ê³„ ({correlation:.2f})")
                    print("     â†’ ê±°ë˜ëŸ‰ ì¦ê°€ ì‹œ ê°€ê²©ë„ ìƒìŠ¹ (ìˆ˜ìš” ì¦ê°€í˜•)")
                elif correlation < -0.5:
                    print(f"  ğŸ“‰ ê°•í•œ ìŒì˜ ìƒê´€ê´€ê³„ ({correlation:.2f})")
                    print("     â†’ ê±°ë˜ëŸ‰ ì¦ê°€ ì‹œ ê°€ê²© í•˜ë½ (ë§¤ë¬¼ ê¸‰ì¦í˜•)")
                else:
                    print(f"  ğŸ“Š ì•½í•œ ìƒê´€ê´€ê³„ ({correlation:.2f})")
                    print("     â†’ ê±°ë˜ëŸ‰ê³¼ ê°€ê²©ì´ ë…ë¦½ì  ì›€ì§ì„")
            
            # ìµœê³ /ìµœì € ì‹œê¸°
            max_price_period = max(periods, key=lambda x: x['mean_price'])
            min_price_period = min(periods, key=lambda x: x['mean_price'])
            max_volume_period = max(periods, key=lambda x: x['volume'])
            min_volume_period = min(periods, key=lambda x: x['volume'])
            
            print(f"  ğŸ”¥ ìµœê³ ê°€ê²©: {max_price_period['period_label']} ({max_price_period['mean_price']:,.0f}ë§Œì›)")
            print(f"  â„ï¸ ìµœì €ê°€ê²©: {min_price_period['period_label']} ({min_price_period['mean_price']:,.0f}ë§Œì›)")
            print(f"  ğŸ“Š ìµœë‹¤ê±°ë˜: {max_volume_period['period_label']} ({max_volume_period['volume']:,}ê±´)")
            print(f"  ğŸ“‰ ìµœì†Œê±°ë˜: {min_volume_period['period_label']} ({min_volume_period['volume']:,}ê±´)")
            
            # ìµœê·¼ íŠ¸ë Œë“œ (ë§ˆì§€ë§‰ 2-3ê¸°ê°„)
            recent_periods = periods[-3:] if len(periods) >= 3 else periods[-2:]
            
            if len(recent_periods) >= 2:
                recent_volume_trend = "ì¦ê°€" if recent_periods[-1]['volume'] > recent_periods[0]['volume'] else "ê°ì†Œ"
                recent_price_trend = "ìƒìŠ¹" if recent_periods[-1]['mean_price'] > recent_periods[0]['mean_price'] else "í•˜ë½"
                
                print(f"  ğŸ“… ìµœê·¼ íŠ¸ë Œë“œ: ê±°ë˜ëŸ‰ {recent_volume_trend}, ê°€ê²© {recent_price_trend}")
                
                # ì½”ë¡œë‚˜ í›„ ì‹œì¥ ë¶„ì„ (2022ë…„ ì´í›„ ë°ì´í„°ê°€ ìˆëŠ” ê²½ìš°)
                post_2022_periods = [p for p in periods if p['start_date'].year >= 2022]
                if len(post_2022_periods) >= 2:
                    price_trend_2022 = "í•˜ë½" if post_2022_periods[-1]['mean_price'] < post_2022_periods[0]['mean_price'] else "ìƒìŠ¹"
                    print(f"  ğŸ¦  2022ë…„ ì´í›„: ê°€ê²© {price_trend_2022} (ê¸ˆë¦¬ì¸ìƒ/ì½”ë¡œë‚˜ íšŒë³µê¸°)")
    
    def run_period_volume_price_analysis(self):
        """ì‹¤ì œ ê¸°ê°„ë³„ ì „ì²´ ë¶„ì„ ì‹¤í–‰"""
        print("ğŸ“Š ì‹¤ì œ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ë™ì‹œ ë¶„ì„ ì‹œì‘")
        print("=" * 60)
        
        self.extract_actual_periods()
        self.create_period_dual_axis_visualization()
        self.create_period_correlation_analysis()
        self.create_comprehensive_period_table()
        self.analyze_period_market_insights()
        
        print(f"\nâœ… ì‹¤ì œ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ë¶„ì„ ì™„ë£Œ")
        print(f"   ì‹¤ì œ ë°ì´í„° ê¸°ê°„ì„ ê¸°ì¤€ìœ¼ë¡œ ì •í™•í•œ ê±°ë˜ëŸ‰-ê°€ê²© ê´€ê³„ë¥¼ ë¶„ì„í–ˆìŠµë‹ˆë‹¤.")
        
        return self.period_data

# ë¶„ì„ ì‹¤í–‰
period_analyzer = PeriodBasedVolumePriceAnalyzer()
period_results = period_analyzer.run_period_volume_price_analysis()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

class PeriodBasedVolumePriceAnalyzer:
    def __init__(self, data_path='data/'):
        self.data_path = data_path
        self.regions = ['ê´‘ì£¼', 'ë¶€ì‚°', 'ì œì£¼']
        self.period_data = {}
        
    def extract_actual_periods(self):
        """ì‹¤ì œ ë°ì´í„° ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ ë° ê°€ê²© ì¶”ì¶œ"""
        print("=== ì‹¤ì œ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ë¶„ì„ ===")
        
        for region in self.regions:
            print(f"\nğŸ“Š [{region}] ì‹¤ì œ ê¸°ê°„ë³„ ë°ì´í„° ì¶”ì¶œ:")
            
            period_analysis = []
            
            for i in range(1, 6):
                try:
                    filename = f"{region}_ì•„íŒŒíŠ¸_{i}.csv"
                    filepath = f"{self.data_path}{filename}"
                    
                    # íŒŒì¼ ë¡œë“œ
                    for encoding in ['utf-8', 'cp949', 'euc-kr']:
                        try:
                            df = pd.read_csv(filepath, encoding=encoding)
                            
                            # ì‹¤ì œ ê¸°ê°„ ë° ê±°ë˜ëŸ‰/ê°€ê²© ë¶„ì„
                            period_info = self.analyze_file_period(df, filename, i)
                            
                            if period_info:
                                period_analysis.append(period_info)
                                
                                print(f"  ğŸ“„ {filename}")
                                print(f"    ğŸ“… ì‹¤ì œê¸°ê°„: {period_info['period_label']}")
                                print(f"    ğŸ“Š ê±°ë˜ëŸ‰: {period_info['volume']:,}ê±´")
                                print(f"    ğŸ’° í‰ê· ê°€ê²©: {period_info['mean_price']:,.0f}ë§Œì›")
                                print(f"    ğŸ’° ì¤‘ìœ„ê°€ê²©: {period_info['median_price']:,.0f}ë§Œì›")
                            
                            break
                        except UnicodeDecodeError:
                            continue
                            
                except Exception as e:
                    print(f"  âŒ {filename}: {e}")
            
            # ì‹¤ì œ ê¸°ê°„ ì‹œì‘ì¼ ê¸°ì¤€ìœ¼ë¡œ ì‹œê°„ìˆœ ì •ë ¬
            if period_analysis:
                period_analysis_sorted = sorted(period_analysis, key=lambda x: x['start_date'])
                
                print(f"\n  ğŸ• ì‹œê°„ìˆœ ì •ë ¬ ê²°ê³¼:")
                for idx, period in enumerate(period_analysis_sorted, 1):
                    print(f"    {idx}ìˆœìœ„: {period['period_label']} (ì›íŒŒì¼: {period['filename']})")
                
                # ì „ê¸° ëŒ€ë¹„ ì¦ê°ë¥  ê³„ì‚°
                for i in range(1, len(period_analysis_sorted)):
                    prev_period = period_analysis_sorted[i-1]
                    curr_period = period_analysis_sorted[i]
                    
                    # ê±°ë˜ëŸ‰ ì¦ê°ë¥ 
                    volume_change = ((curr_period['volume'] - prev_period['volume']) / prev_period['volume']) * 100
                    curr_period['volume_change'] = volume_change
                    
                    # ê°€ê²© ì¦ê°ë¥ 
                    price_change = ((curr_period['mean_price'] - prev_period['mean_price']) / prev_period['mean_price']) * 100
                    curr_period['price_change'] = price_change
                    
                    median_change = ((curr_period['median_price'] - prev_period['median_price']) / prev_period['median_price']) * 100
                    curr_period['median_change'] = median_change
                
                # ì²« ë²ˆì§¸ ê¸°ê°„ì€ ì¦ê°ë¥  ì—†ìŒ
                period_analysis_sorted[0]['volume_change'] = None
                period_analysis_sorted[0]['price_change'] = None
                period_analysis_sorted[0]['median_change'] = None
                
                self.period_data[region] = period_analysis_sorted
    
    def analyze_file_period(self, df, filename, file_num):
        """ê°œë³„ íŒŒì¼ì˜ ì‹¤ì œ ê¸°ê°„ ë° ê±°ë˜ëŸ‰/ê°€ê²© ë¶„ì„"""
        # ë‚ ì§œ ì»¬ëŸ¼ ì°¾ê¸° ë° íŒŒì‹±
        date_cols = [col for col in df.columns if any(keyword in col.lower() 
                    for keyword in ['ë‚ ì§œ', 'date', 'ê³„ì•½', 'ê±°ë˜', 'ë…„ì›”'])]
        
        parsed_dates = None
        for date_col in date_cols:
            try:
                formats = ['%Y-%m-%d', '%Y.%m.%d', '%Y/%m/%d', '%Y-%m', '%Y.%m', '%Y%m']
                
                for fmt in formats:
                    try:
                        temp_dates = pd.to_datetime(df[date_col], format=fmt, errors='coerce')
                        if temp_dates.notna().sum() / len(temp_dates) > 0.7:  # 70% ì´ìƒ ì„±ê³µ
                            parsed_dates = temp_dates
                            break
                    except:
                        continue
                
                if parsed_dates is None:
                    parsed_dates = pd.to_datetime(df[date_col], errors='coerce')
                
                if parsed_dates.notna().sum() > len(df) * 0.5:  # 50% ì´ìƒ ì„±ê³µ
                    break
            except:
                continue
        
        # ê°€ê²© ì»¬ëŸ¼ ì°¾ê¸° ë° íŒŒì‹±
        price_keywords = ['ê°€ê²©', 'ê¸ˆì•¡', 'price', 'ê±°ë˜ê¸ˆì•¡', 'ë§¤ë§¤ê°€']
        price_cols = [col for col in df.columns if any(keyword in col.lower() for keyword in price_keywords)]
        
        if not price_cols or parsed_dates is None:
            return None
        
        main_price_col = price_cols[0]
        
        try:
            if df[main_price_col].dtype == 'object':
                price_clean = df[main_price_col].astype(str)
                price_clean = price_clean.str.replace(',', '').str.replace('ë§Œì›', '').str.replace('ì›', '')
                price_clean = price_clean.str.extract('(\d+\.?\d*)')[0]
                cleaned_prices = pd.to_numeric(price_clean, errors='coerce')
            else:
                cleaned_prices = pd.to_numeric(df[main_price_col], errors='coerce')
        except:
            return None
        
        # ìœ íš¨í•œ ë°ì´í„°ë§Œ í•„í„°ë§
        df['ê±°ë˜ë‚ ì§œ'] = parsed_dates
        df['ê±°ë˜ê°€ê²©'] = cleaned_prices
        
        valid_data = df[(df['ê±°ë˜ë‚ ì§œ'].notna()) & 
                       (df['ê±°ë˜ê°€ê²©'].notna()) & 
                       (df['ê±°ë˜ê°€ê²©'] > 0) & 
                       (df['ê±°ë˜ê°€ê²©'] < 1000000)]
        
        if len(valid_data) < 50:  # ìµœì†Œ 50ê±´ ì´ìƒ
            return None
        
        # ì‹¤ì œ ê¸°ê°„ ê³„ì‚°
        start_date = valid_data['ê±°ë˜ë‚ ì§œ'].min()
        end_date = valid_data['ê±°ë˜ë‚ ì§œ'].max()
        
        # ê¸°ê°„ ë¼ë²¨ ìƒì„± (ë…„.ì›” í˜•íƒœ)
        start_ym = f"{start_date.year}.{start_date.month}"
        end_ym = f"{end_date.year}.{end_date.month}"
        period_label = f"{start_ym}~{end_ym}"
        
        # ê±°ë˜ëŸ‰ ë° ê°€ê²© í†µê³„
        volume = len(valid_data)
        mean_price = valid_data['ê±°ë˜ê°€ê²©'].mean()
        median_price = valid_data['ê±°ë˜ê°€ê²©'].median()
        price_std = valid_data['ê±°ë˜ê°€ê²©'].std()
        
        return {
            'filename': filename,
            'file_num': file_num,
            'start_date': start_date,
            'end_date': end_date,
            'period_label': period_label,
            'start_ym': start_ym,
            'end_ym': end_ym,
            'volume': volume,
            'mean_price': mean_price,
            'median_price': median_price,
            'price_std': price_std,
            'duration_days': (end_date - start_date).days
        }
    
    def create_period_dual_axis_visualization(self):
        """ì‹¤ì œ ê¸°ê°„ë³„ ì´ì¤‘ì¶• ì‹œê°í™”"""
        print("\n=== ì‹¤ì œ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ì´ì¤‘ì¶• ì‹œê°í™” ===")
        
        # ì§€ì—­ë³„ ì„œë¸Œí”Œë¡¯
        fig = make_subplots(
            rows=len(self.regions), cols=1,
            subplot_titles=[f'{region} - ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs í‰ê· ê°€ê²© (ì‹¤ì œ ê¸°ê°„ ê¸°ì¤€)' for region in self.regions],
            specs=[[{"secondary_y": True}] for _ in self.regions],
            vertical_spacing=0.15
        )
        
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for row, region in enumerate(self.regions, 1):
            if region in self.period_data:
                periods = self.period_data[region]
                
                x_labels = [f"{i+1}ê¸°\n({p['period_label']})" for i, p in enumerate(periods)]
                volumes = [p['volume'] for p in periods]
                prices = [p['mean_price'] for p in periods]
                period_labels = [p['period_label'] for p in periods]
                
                # ê±°ë˜ëŸ‰ (ë§‰ëŒ€ê·¸ë˜í”„)
                fig.add_trace(
                    go.Bar(
                        x=list(range(len(periods))),
                        y=volumes,
                        name=f'{region} ê±°ë˜ëŸ‰',
                        marker_color=colors[region],
                        opacity=0.7,
                        text=[f'{v:,}ê±´' for v in volumes],
                        textposition='auto',
                        hovertemplate=f'<b>{region} ê±°ë˜ëŸ‰</b><br>ê¸°ê°„: %{{customdata}}<br>ê±°ë˜ëŸ‰: %{{y:,}}ê±´<extra></extra>',
                        customdata=period_labels,
                        showlegend=(row==1)
                    ),
                    row=row, col=1,
                    secondary_y=False
                )
                
                # í‰ê· ê°€ê²© (ì„ ê·¸ë˜í”„)
                fig.add_trace(
                    go.Scatter(
                        x=list(range(len(periods))),
                        y=prices,
                        mode='lines+markers+text',
                        name=f'{region} í‰ê· ê°€ê²©',
                        line=dict(color=colors[region], width=4),
                        marker=dict(size=12, color='white', line=dict(color=colors[region], width=3)),
                        text=[f'{p:,.0f}ë§Œì›' for p in prices],
                        textposition='middle right',  # í…ìŠ¤íŠ¸ ìœ„ì¹˜ë¥¼ ì˜¤ë¥¸ìª½ìœ¼ë¡œ ë³€ê²½
                        textfont=dict(color=colors[region], size=11, family="Arial Black"),
                        hovertemplate=f'<b>{region} í‰ê· ê°€ê²©</b><br>ê¸°ê°„: %{{customdata}}<br>í‰ê· ê°€ê²©: %{{y:,.0f}}ë§Œì›<extra></extra>',
                        customdata=period_labels,
                        showlegend=(row==1)
                    ),
                    row=row, col=1,
                    secondary_y=True
                )
                
                # xì¶• ë¼ë²¨ ì„¤ì • (ê¸°ê°„ í‘œì‹œ)
                fig.update_xaxes(
                    tickmode='array',
                    tickvals=list(range(len(periods))),
                    ticktext=x_labels,
                    row=row, col=1
                )
        
        # ì¶• ë¼ë²¨ ì„¤ì •
        for i in range(1, len(self.regions) + 1):
            fig.update_yaxes(title_text="ê±°ë˜ëŸ‰ (ê±´)", secondary_y=False, row=i, col=1)
            fig.update_yaxes(title_text="í‰ê· ê°€ê²© (ë§Œì›)", secondary_y=True, row=i, col=1)
        
        fig.update_layout(
            title='ğŸ“Š ì‹¤ì œ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs í‰ê· ê°€ê²© ë™ì‹œ ë¶„ì„',
            title_x=0.5,
            title_font_size=20,
            height=350 * len(self.regions),
            margin=dict(t=120, b=100, l=80, r=120)  # ì˜¤ë¥¸ìª½, ìœ„ìª½ ì—¬ë°± ì¶©ë¶„íˆ í™•ë³´
        )
        
        fig.show()
    
    def create_period_correlation_analysis(self):
        """ì‹¤ì œ ê¸°ê°„ë³„ ìƒê´€ê´€ê³„ ë¶„ì„"""
        print("\n=== ê¸°ê°„ë³„ ê±°ë˜ëŸ‰-ê°€ê²© ìƒê´€ê´€ê³„ ë¶„ì„ ===")
        
        # ì‚°ì ë„
        fig1 = go.Figure()
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for region in self.regions:
            if region in self.period_data:
                periods = self.period_data[region]
                
                volumes = [p['volume'] for p in periods]
                prices = [p['mean_price'] for p in periods]
                period_labels = [p['period_label'] for p in periods]
                
                # ìƒê´€ê³„ìˆ˜ ê³„ì‚°
                if len(volumes) > 2:
                    correlation = np.corrcoef(volumes, prices)[0, 1]
                    
                    fig1.add_trace(go.Scatter(
                        x=volumes,
                        y=prices,
                        mode='markers+text',
                        name=f'{region} (ìƒê´€ê³„ìˆ˜: {correlation:.2f})',
                        marker=dict(size=15, color=colors[region]),
                        text=period_labels,
                        textposition='top center',
                        textfont=dict(size=10),
                        hovertemplate=f'<b>{region}</b><br>ê¸°ê°„: %{{text}}<br>ê±°ë˜ëŸ‰: %{{x:,}}ê±´<br>í‰ê· ê°€ê²©: %{{y:,.0f}}ë§Œì›<extra></extra>'
                    ))
        
        fig1.update_layout(
            title='ğŸ“ˆ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs í‰ê· ê°€ê²© ìƒê´€ê´€ê³„ (ì‚°ì ë„)',
            title_x=0.5,
            title_font_size=18,
            xaxis_title='ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ (ê±´)',
            yaxis_title='í‰ê· ê°€ê²© (ë§Œì›)',
            height=500
        )
        
        fig1.show()
        
        # ì¦ê°ë¥  ë¹„êµ
        self.create_period_change_rate_comparison()
    
    def create_period_change_rate_comparison(self):
        """ê¸°ê°„ë³„ ì¦ê°ë¥  ë¹„êµ"""
        fig = make_subplots(
            rows=1, cols=2,
            subplot_titles=('ê±°ë˜ëŸ‰ ì¦ê°ë¥  (ì „ê¸° ëŒ€ë¹„)', 'í‰ê· ê°€ê²© ì¦ê°ë¥  (ì „ê¸° ëŒ€ë¹„)'),
            specs=[[{"secondary_y": False}, {"secondary_y": False}]]
        )
        
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for region in self.regions:
            if region in self.period_data:
                periods = self.period_data[region]
                
                # ì¦ê°ë¥ ì´ ìˆëŠ” ë°ì´í„°ë§Œ (2ê¸°ë¶€í„°)
                change_data = [p for p in periods if p['volume_change'] is not None and p['price_change'] is not None]
                
                if change_data:
                    x_labels = [f"{periods.index(p)+1}ê¸°\n({p['period_label']})" for p in change_data]
                    volume_changes = [p['volume_change'] for p in change_data]
                    price_changes = [p['price_change'] for p in change_data]
                    
                    # ê±°ë˜ëŸ‰ ì¦ê°ë¥ 
                    fig.add_trace(
                        go.Bar(
                            x=x_labels,
                            y=volume_changes,
                            name=f'{region}',
                            marker_color=colors[region],
                            text=[f'{v:+.1f}%' for v in volume_changes],
                            textposition='auto',
                            showlegend=True
                        ),
                        row=1, col=1
                    )
                    
                    # ê°€ê²© ì¦ê°ë¥ 
                    fig.add_trace(
                        go.Bar(
                            x=x_labels,
                            y=price_changes,
                            name=f'{region}',
                            marker_color=colors[region],
                            text=[f'{v:+.1f}%' for v in price_changes],
                            textposition='auto',
                            showlegend=False
                        ),
                        row=1, col=2
                    )
        
        # 0% ê¸°ì¤€ì„ 
        fig.add_hline(y=0, line=dict(color='gray', dash='dash', width=1), row=1, col=1)
        fig.add_hline(y=0, line=dict(color='gray', dash='dash', width=1), row=1, col=2)
        
        fig.update_layout(
            title='ğŸ“Š ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ì¦ê°ë¥  ë¹„êµ (ì „ê¸° ëŒ€ë¹„)',
            title_x=0.5,
            title_font_size=18,
            height=500
        )
        
        fig.update_yaxes(title_text="ê±°ë˜ëŸ‰ ì¦ê°ë¥  (%)", row=1, col=1)
        fig.update_yaxes(title_text="ê°€ê²© ì¦ê°ë¥  (%)", row=1, col=2)
        
        fig.show()
    
    def create_comprehensive_period_table(self):
        """ì‹¤ì œ ê¸°ê°„ë³„ ì¢…í•© í…Œì´ë¸”"""
        print("\n=== ì‹¤ì œ ê¸°ê°„ë³„ ì¢…í•© ìš”ì•½ í…Œì´ë¸” ===")
        
        table_data = []
        
        for region in self.regions:
            if region in self.period_data:
                for i, period in enumerate(self.period_data[region]):
                    vol_change_str = f"{period['volume_change']:+.1f}%" if period['volume_change'] is not None else "-"
                    price_change_str = f"{period['price_change']:+.1f}%" if period['price_change'] is not None else "-"
                    median_change_str = f"{period['median_change']:+.1f}%" if period['median_change'] is not None else "-"
                    
                    table_data.append([
                        region,
                        f"{i+1}ê¸°",
                        period['period_label'],
                        f"{period['duration_days']}ì¼",
                        f"{period['volume']:,}ê±´",
                        vol_change_str,
                        f"{period['mean_price']:,.0f}ë§Œì›",
                        price_change_str,
                        f"{period['median_price']:,.0f}ë§Œì›",
                        period['filename']
                    ])
        
        if table_data:
            fig = go.Figure(data=[go.Table(
                header=dict(
                    values=['ì§€ì—­', 'ìˆœì„œ', 'ì‹¤ì œ ê¸°ê°„', 'ê¸°ê°„', 'ê±°ë˜ëŸ‰', 'ê±°ë˜ëŸ‰ì¦ê°ë¥ ', 'í‰ê· ê°€ê²©', 'ê°€ê²©ì¦ê°ë¥ ', 'ì¤‘ìœ„ê°€ê²©', 'ì›ë³¸íŒŒì¼'],
                    fill_color='lightblue',
                    align='center',
                    font=dict(size=11, color='black'),
                    height=45
                ),
                cells=dict(
                    values=list(zip(*table_data)),
                    fill_color='white',
                    align='center',
                    font=dict(size=10),
                    height=35
                )
            )])
            
            fig.update_layout(
                title='ğŸ“‹ ì‹¤ì œ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ì¢…í•© ë¶„ì„í‘œ',
                title_x=0.5,
                title_font_size=16,
                height=600,
                margin=dict(l=20, r=20, t=80, b=20)
            )
            
            fig.show()
    
    def analyze_period_market_insights(self):
        """ì‹¤ì œ ê¸°ê°„ë³„ ì‹œì¥ ì¸ì‚¬ì´íŠ¸"""
        print("\n=== ğŸ” ì‹¤ì œ ê¸°ê°„ë³„ ì‹œì¥ ë¶„ì„ ì¸ì‚¬ì´íŠ¸ ===")
        
        for region in self.regions:
            if region not in self.period_data:
                continue
                
            periods = self.period_data[region]
            print(f"\nğŸ  [{region}] ê¸°ê°„ë³„ ì‹œì¥ íŠ¹ì„±:")
            
            if len(periods) < 2:
                print("  âš ï¸ ë¶„ì„ ê¸°ê°„ ë¶€ì¡±")
                continue
            
            # ê±°ë˜ëŸ‰ê³¼ ê°€ê²©ì˜ ìƒê´€ê´€ê³„
            volumes = [p['volume'] for p in periods]
            prices = [p['mean_price'] for p in periods]
            
            if len(volumes) > 2:
                correlation = np.corrcoef(volumes, prices)[0, 1]
                
                if correlation > 0.5:
                    print(f"  ğŸ“ˆ ê°•í•œ ì–‘ì˜ ìƒê´€ê´€ê³„ ({correlation:.2f})")
                    print("     â†’ ê±°ë˜ëŸ‰ ì¦ê°€ ì‹œ ê°€ê²©ë„ ìƒìŠ¹ (ìˆ˜ìš” ì¦ê°€í˜•)")
                elif correlation < -0.5:
                    print(f"  ğŸ“‰ ê°•í•œ ìŒì˜ ìƒê´€ê´€ê³„ ({correlation:.2f})")
                    print("     â†’ ê±°ë˜ëŸ‰ ì¦ê°€ ì‹œ ê°€ê²© í•˜ë½ (ë§¤ë¬¼ ê¸‰ì¦í˜•)")
                else:
                    print(f"  ğŸ“Š ì•½í•œ ìƒê´€ê´€ê³„ ({correlation:.2f})")
                    print("     â†’ ê±°ë˜ëŸ‰ê³¼ ê°€ê²©ì´ ë…ë¦½ì  ì›€ì§ì„")
            
            # ìµœê³ /ìµœì € ì‹œê¸°
            max_price_period = max(periods, key=lambda x: x['mean_price'])
            min_price_period = min(periods, key=lambda x: x['mean_price'])
            max_volume_period = max(periods, key=lambda x: x['volume'])
            min_volume_period = min(periods, key=lambda x: x['volume'])
            
            print(f"  ğŸ”¥ ìµœê³ ê°€ê²©: {max_price_period['period_label']} ({max_price_period['mean_price']:,.0f}ë§Œì›)")
            print(f"  â„ï¸ ìµœì €ê°€ê²©: {min_price_period['period_label']} ({min_price_period['mean_price']:,.0f}ë§Œì›)")
            print(f"  ğŸ“Š ìµœë‹¤ê±°ë˜: {max_volume_period['period_label']} ({max_volume_period['volume']:,}ê±´)")
            print(f"  ğŸ“‰ ìµœì†Œê±°ë˜: {min_volume_period['period_label']} ({min_volume_period['volume']:,}ê±´)")
            
            # ìµœê·¼ íŠ¸ë Œë“œ (ë§ˆì§€ë§‰ 2-3ê¸°ê°„)
            recent_periods = periods[-3:] if len(periods) >= 3 else periods[-2:]
            
            if len(recent_periods) >= 2:
                recent_volume_trend = "ì¦ê°€" if recent_periods[-1]['volume'] > recent_periods[0]['volume'] else "ê°ì†Œ"
                recent_price_trend = "ìƒìŠ¹" if recent_periods[-1]['mean_price'] > recent_periods[0]['mean_price'] else "í•˜ë½"
                
                print(f"  ğŸ“… ìµœê·¼ íŠ¸ë Œë“œ: ê±°ë˜ëŸ‰ {recent_volume_trend}, ê°€ê²© {recent_price_trend}")
                
                # ì½”ë¡œë‚˜ í›„ ì‹œì¥ ë¶„ì„ (2022ë…„ ì´í›„ ë°ì´í„°ê°€ ìˆëŠ” ê²½ìš°)
                post_2022_periods = [p for p in periods if p['start_date'].year >= 2022]
                if len(post_2022_periods) >= 2:
                    price_trend_2022 = "í•˜ë½" if post_2022_periods[-1]['mean_price'] < post_2022_periods[0]['mean_price'] else "ìƒìŠ¹"
                    print(f"  ğŸ¦  2022ë…„ ì´í›„: ê°€ê²© {price_trend_2022} (ê¸ˆë¦¬ì¸ìƒ/ì½”ë¡œë‚˜ íšŒë³µê¸°)")
    
    def run_period_volume_price_analysis(self):
        """ì‹¤ì œ ê¸°ê°„ë³„ ì „ì²´ ë¶„ì„ ì‹¤í–‰"""
        print("ğŸ“Š ì‹¤ì œ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ë™ì‹œ ë¶„ì„ ì‹œì‘")
        print("=" * 60)
        
        self.extract_actual_periods()
        self.create_period_dual_axis_visualization()
        self.create_period_correlation_analysis()
        self.create_comprehensive_period_table()
        self.analyze_period_market_insights()
        
        print(f"\nâœ… ì‹¤ì œ ê¸°ê°„ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ë¶„ì„ ì™„ë£Œ")
        print(f"   ì‹¤ì œ ë°ì´í„° ê¸°ê°„ì„ ê¸°ì¤€ìœ¼ë¡œ ì •í™•í•œ ê±°ë˜ëŸ‰-ê°€ê²© ê´€ê³„ë¥¼ ë¶„ì„í–ˆìŠµë‹ˆë‹¤.")
        
        return self.period_data

# ë¶„ì„ ì‹¤í–‰
period_analyzer = PeriodBasedVolumePriceAnalyzer()
period_results = period_analyzer.run_period_volume_price_analysis()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

class AreaBasedAnalyzer:
    def __init__(self, data_path='data/'):
        self.data_path = data_path
        self.regions = ['ê´‘ì£¼', 'ë¶€ì‚°', 'ì œì£¼']
        self.area_data = {}
        
        # ë©´ì ë³„ ë¶„ë¥˜ ê¸°ì¤€ (ã¡)
        self.area_categories = {
            'ì†Œí˜•': (0, 60),      # ~60ã¡ (ì•½ ~18í‰)
            'ì¤‘í˜•': (60, 85),     # 60~85ã¡ (ì•½ 18~26í‰)
            'ëŒ€í˜•': (85, 200)     # 85ã¡~ (ì•½ 26í‰~)
        }
        
    def load_and_classify_by_area(self):
        """ë©´ì ë³„ ë¶„ë¥˜ ë° ë°ì´í„° ë¡œë“œ"""
        print("=== ì „ìš©ë©´ì ë³„ ë¶„ë¥˜ ë° ë¶„ì„ ===")
        
        for region in self.regions:
            print(f"\nğŸ  [{region}] ë©´ì ë³„ ë°ì´í„° ë¶„ë¥˜:")
            
            all_data = []
            
            for i in range(1, 6):
                try:
                    filename = f"{region}_ì•„íŒŒíŠ¸_{i}.csv"
                    filepath = f"{self.data_path}{filename}"
                    
                    # íŒŒì¼ ë¡œë“œ
                    for encoding in ['utf-8', 'cp949', 'euc-kr']:
                        try:
                            df = pd.read_csv(filepath, encoding=encoding)
                            
                            # ë‚ ì§œ, ê°€ê²©, ë©´ì  ì²˜ë¦¬
                            processed_df = self.process_area_price_data(df, filename, i)
                            
                            if processed_df is not None and len(processed_df) > 0:
                                all_data.append(processed_df)
                                print(f"  âœ… {filename}: {len(processed_df)}ê±´ ì²˜ë¦¬ì™„ë£Œ")
                            
                            break
                        except UnicodeDecodeError:
                            continue
                            
                except Exception as e:
                    print(f"  âŒ {filename}: {e}")
            
            if all_data:
                # ëª¨ë“  ë°ì´í„° í†µí•©
                combined_df = pd.concat(all_data, ignore_index=True)
                
                # ë©´ì ë³„ ë¶„ë¥˜
                area_analysis = self.classify_and_analyze_by_area(combined_df, region)
                self.area_data[region] = area_analysis
                
                print(f"  ğŸ“Š ì´ {len(combined_df):,}ê±´ â†’ ë©´ì ë³„ ë¶„ë¥˜ ì™„ë£Œ")
    
    def process_area_price_data(self, df, filename, file_num):
        """ë‚ ì§œ, ê°€ê²©, ë©´ì  ë°ì´í„° ì²˜ë¦¬"""
        # ë‚ ì§œ ì»¬ëŸ¼ ì²˜ë¦¬
        date_cols = [col for col in df.columns if any(keyword in col.lower() 
                    for keyword in ['ë‚ ì§œ', 'date', 'ê³„ì•½', 'ê±°ë˜', 'ë…„ì›”'])]
        
        parsed_dates = None
        for date_col in date_cols:
            try:
                formats = ['%Y-%m-%d', '%Y.%m.%d', '%Y/%m/%d', '%Y-%m', '%Y.%m', '%Y%m']
                
                for fmt in formats:
                    try:
                        temp_dates = pd.to_datetime(df[date_col], format=fmt, errors='coerce')
                        if temp_dates.notna().sum() / len(temp_dates) > 0.7:
                            parsed_dates = temp_dates
                            break
                    except:
                        continue
                
                if parsed_dates is None:
                    parsed_dates = pd.to_datetime(df[date_col], errors='coerce')
                
                if parsed_dates.notna().sum() > len(df) * 0.5:
                    break
            except:
                continue
        
        # ê°€ê²© ì»¬ëŸ¼ ì²˜ë¦¬
        price_keywords = ['ê°€ê²©', 'ê¸ˆì•¡', 'price', 'ê±°ë˜ê¸ˆì•¡', 'ë§¤ë§¤ê°€']
        price_cols = [col for col in df.columns if any(keyword in col.lower() for keyword in price_keywords)]
        
        # ë©´ì  ì»¬ëŸ¼ ì²˜ë¦¬
        area_keywords = ['ë©´ì ', 'area', 'í‰', 'ã¡', 'm2', 'ì „ìš©ë©´ì ']
        area_cols = [col for col in df.columns if any(keyword in col.lower() for keyword in area_keywords)]
        
        if not price_cols or not area_cols or parsed_dates is None:
            return None
        
        main_price_col = price_cols[0]
        main_area_col = area_cols[0]
        
        try:
            # ê°€ê²© ì²˜ë¦¬
            if df[main_price_col].dtype == 'object':
                price_clean = df[main_price_col].astype(str)
                price_clean = price_clean.str.replace(',', '').str.replace('ë§Œì›', '').str.replace('ì›', '')
                price_clean = price_clean.str.extract('(\d+\.?\d*)')[0]
                cleaned_prices = pd.to_numeric(price_clean, errors='coerce')
            else:
                cleaned_prices = pd.to_numeric(df[main_price_col], errors='coerce')
            
            # ë©´ì  ì²˜ë¦¬
            if df[main_area_col].dtype == 'object':
                area_clean = df[main_area_col].astype(str)
                area_clean = area_clean.str.replace(',', '').str.extract('(\d+\.?\d*)')[0]
                cleaned_areas = pd.to_numeric(area_clean, errors='coerce')
            else:
                cleaned_areas = pd.to_numeric(df[main_area_col], errors='coerce')
        except:
            return None
        
        # ìœ íš¨í•œ ë°ì´í„°ë§Œ ì¶”ì¶œ
        df['ê±°ë˜ë‚ ì§œ'] = parsed_dates
        df['ê±°ë˜ê°€ê²©'] = cleaned_prices
        df['ì „ìš©ë©´ì '] = cleaned_areas
        df['í‰ìˆ˜'] = df['ì „ìš©ë©´ì '] / 3.3058
        df['íŒŒì¼ë²ˆí˜¸'] = file_num
        
        # ì‹¤ì œ ê¸°ê°„ ê³„ì‚°
        valid_dates = df['ê±°ë˜ë‚ ì§œ'].dropna()
        if len(valid_dates) > 0:
            start_date = valid_dates.min()
            end_date = valid_dates.max()
            df['ì‹¤ì œê¸°ê°„'] = f"{start_date.strftime('%Y.%m')}~{end_date.strftime('%Y.%m')}"
        
        valid_data = df[(df['ê±°ë˜ë‚ ì§œ'].notna()) & 
                       (df['ê±°ë˜ê°€ê²©'].notna()) & 
                       (df['ì „ìš©ë©´ì '].notna()) &
                       (df['ê±°ë˜ê°€ê²©'] > 0) & 
                       (df['ê±°ë˜ê°€ê²©'] < 1000000) &
                       (df['ì „ìš©ë©´ì '] > 10) &  # ìµœì†Œ 10ã¡ ì´ìƒ
                       (df['ì „ìš©ë©´ì '] < 300)]  # ìµœëŒ€ 300ã¡ ë¯¸ë§Œ
        
        return valid_data
    
    def classify_and_analyze_by_area(self, df, region):
        """ë©´ì ë³„ ë¶„ë¥˜ ë° ë¶„ì„"""
        print(f"\n  ğŸ“ [{region}] ë©´ì ë³„ ë¶„ë¥˜:")
        
        # ë©´ì  ì¹´í…Œê³ ë¦¬ ë¶„ë¥˜
        def get_area_category(area):
            for category, (min_area, max_area) in self.area_categories.items():
                if min_area <= area < max_area:
                    return category
            return 'ëŒ€í˜•'  # 85ã¡ ì´ìƒì€ ëª¨ë‘ ëŒ€í˜•
        
        df['ë©´ì êµ¬ë¶„'] = df['ì „ìš©ë©´ì '].apply(get_area_category)
        
        # ë©´ì ë³„ í†µê³„
        area_stats = df.groupby('ë©´ì êµ¬ë¶„').agg({
            'ê±°ë˜ê°€ê²©': ['count', 'mean', 'median'],
            'ì „ìš©ë©´ì ': 'mean',
            'í‰ìˆ˜': 'mean'
        }).round(2)
        
        area_stats.columns = ['ê±°ë˜ê±´ìˆ˜', 'í‰ê· ê°€ê²©', 'ì¤‘ìœ„ê°€ê²©', 'í‰ê· ë©´ì ', 'í‰ê· í‰ìˆ˜']
        
        print(f"    ë©´ì ë³„ ê±°ë˜ í˜„í™©:")
        for category in ['ì†Œí˜•', 'ì¤‘í˜•', 'ëŒ€í˜•']:
            if category in area_stats.index:
                stats = area_stats.loc[category]
                total_count = len(df)
                ratio = (stats['ê±°ë˜ê±´ìˆ˜'] / total_count) * 100
                
                print(f"      {category} ({self.area_categories.get(category, (85, 200))[0]}~{self.area_categories.get(category, (85, 200))[1]}ã¡): "
                      f"{stats['ê±°ë˜ê±´ìˆ˜']:,}ê±´ ({ratio:.1f}%) - "
                      f"í‰ê·  {stats['í‰ê· ê°€ê²©']:,.0f}ë§Œì›")
        
        # ê¸°ê°„ë³„ ë©´ì ë³„ ë¶„ì„
        period_area_analysis = self.analyze_period_area_trends(df, region)
        
        return {
            'raw_data': df,
            'area_stats': area_stats,
            'period_analysis': period_area_analysis
        }
    
    def analyze_period_area_trends(self, df, region):
        """ê¸°ê°„ë³„ ë©´ì ë³„ íŠ¸ë Œë“œ ë¶„ì„"""
        # íŒŒì¼ë³„(ê¸°ê°„ë³„) + ë©´ì ë³„ ì§‘ê³„
        period_area = df.groupby(['ì‹¤ì œê¸°ê°„', 'ë©´ì êµ¬ë¶„']).agg({
            'ê±°ë˜ê°€ê²©': ['count', 'mean'],
            'íŒŒì¼ë²ˆí˜¸': 'first'
        }).round(0)
        
        period_area.columns = ['ê±°ë˜ê±´ìˆ˜', 'í‰ê· ê°€ê²©', 'íŒŒì¼ë²ˆí˜¸']
        period_area = period_area.reset_index()
        
        # ê¸°ê°„ë³„ ì •ë ¬ (íŒŒì¼ë²ˆí˜¸ ê¸°ì¤€)
        period_area = period_area.sort_values(['íŒŒì¼ë²ˆí˜¸', 'ë©´ì êµ¬ë¶„'])
        
        return period_area
    
    def create_area_price_trend_chart(self):
        """ì „ìš©ë©´ì ë³„ í‰ê· ê°€ê²© ì¶”ì´ ì°¨íŠ¸"""
        print("\n=== ì „ìš©ë©´ì ë³„ í‰ê· ê°€ê²© ì¶”ì´ ì°¨íŠ¸ ===")
        
        fig = make_subplots(
            rows=len(self.regions), cols=1,
            subplot_titles=[f'{region} - ë©´ì ë³„ í‰ê· ê°€ê²© ì¶”ì´ (5ë…„ê°„)' for region in self.regions],
            vertical_spacing=0.12
        )
        
        # ë©´ì ë³„ ìƒ‰ìƒ
        area_colors = {
            'ì†Œí˜•': '#FF6B6B',  # ë¹¨ê°• (ì†Œí˜•)
            'ì¤‘í˜•': '#4ECDC4',  # ì²­ë¡ (ì¤‘í˜•)
            'ëŒ€í˜•': '#45B7D1'   # íŒŒë‘ (ëŒ€í˜•)
        }
        
        for row, region in enumerate(self.regions, 1):
            if region not in self.area_data:
                continue
                
            period_analysis = self.area_data[region]['period_analysis']
            
            # ê¸°ê°„ë³„ë¡œ ì •ë¦¬
            periods = sorted(period_analysis['ì‹¤ì œê¸°ê°„'].unique(), 
                           key=lambda x: period_analysis[period_analysis['ì‹¤ì œê¸°ê°„']==x]['íŒŒì¼ë²ˆí˜¸'].iloc[0])
            
            for category in ['ì†Œí˜•', 'ì¤‘í˜•', 'ëŒ€í˜•']:
                category_data = period_analysis[period_analysis['ë©´ì êµ¬ë¶„'] == category]
                
                if not category_data.empty:
                    x_values = []
                    y_values = []
                    
                    for period in periods:
                        period_data = category_data[category_data['ì‹¤ì œê¸°ê°„'] == period]
                        if not period_data.empty:
                            x_values.append(period)
                            y_values.append(period_data['í‰ê· ê°€ê²©'].iloc[0])
                        else:
                            x_values.append(period)
                            y_values.append(None)
                    
                    # None ê°’ ì œê±°
                    valid_data = [(x, y) for x, y in zip(x_values, y_values) if y is not None]
                    if valid_data:
                        x_clean, y_clean = zip(*valid_data)
                        
                        fig.add_trace(
                            go.Scatter(
                                x=x_clean,
                                y=y_clean,
                                mode='lines+markers+text',
                                name=f'{category}',
                                line=dict(color=area_colors[category], width=3),
                                marker=dict(size=8),
                                text=[f'{y:,.0f}ë§Œì›' for y in y_clean],
                                textposition='top center',
                                textfont=dict(size=10, color=area_colors[category]),
                                hovertemplate=f'<b>{region} {category}</b><br>ê¸°ê°„: %{{x}}<br>í‰ê· ê°€ê²©: %{{y:,.0f}}ë§Œì›<extra></extra>',
                                showlegend=(row==1)
                            ),
                            row=row, col=1
                        )
        
        fig.update_layout(
            title='ğŸ  ì „ìš©ë©´ì ë³„ í‰ê· ê°€ê²© ì¶”ì´ (5ë…„ê°„ ë¼ì¸ì°¨íŠ¸)',
            title_x=0.5,
            title_font_size=20,
            height=350 * len(self.regions),
            margin=dict(t=120, b=100, l=80, r=120)
        )
        
        fig.show()
    
    def create_area_volume_price_chart(self):
        """ê·œëª¨ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ë™ì‹œ ê·¸ë˜í”„"""
        print("\n=== ê·œëª¨ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ë™ì‹œ ê·¸ë˜í”„ ===")
        
        fig = make_subplots(
            rows=len(self.regions), cols=1,
            subplot_titles=[f'{region} - ë©´ì ë³„ ê±°ë˜ëŸ‰ vs í‰ê· ê°€ê²©' for region in self.regions],
            specs=[[{"secondary_y": True}] for _ in self.regions],
            vertical_spacing=0.15
        )
        
        area_colors = {
            'ì†Œí˜•': '#FF6B6B',
            'ì¤‘í˜•': '#4ECDC4', 
            'ëŒ€í˜•': '#45B7D1'
        }
        
        for row, region in enumerate(self.regions, 1):
            if region not in self.area_data:
                continue
                
            area_stats = self.area_data[region]['area_stats']
            
            categories = ['ì†Œí˜•', 'ì¤‘í˜•', 'ëŒ€í˜•']
            volumes = []
            prices = []
            
            for category in categories:
                if category in area_stats.index:
                    volumes.append(area_stats.loc[category, 'ê±°ë˜ê±´ìˆ˜'])
                    prices.append(area_stats.loc[category, 'í‰ê· ê°€ê²©'])
                else:
                    volumes.append(0)
                    prices.append(0)
            
            # ê±°ë˜ëŸ‰ (ë§‰ëŒ€ê·¸ë˜í”„)
            fig.add_trace(
                go.Bar(
                    x=categories,
                    y=volumes,
                    name=f'{region} ê±°ë˜ëŸ‰',
                    marker_color=[area_colors[cat] for cat in categories],
                    opacity=0.7,
                    text=[f'{v:,}ê±´' for v in volumes],
                    textposition='auto',
                    hovertemplate=f'<b>{region} ê±°ë˜ëŸ‰</b><br>ë©´ì êµ¬ë¶„: %{{x}}<br>ê±°ë˜ëŸ‰: %{{y:,}}ê±´<extra></extra>',
                    showlegend=(row==1)
                ),
                row=row, col=1,
                secondary_y=False
            )
            
            # í‰ê· ê°€ê²© (ì„ ê·¸ë˜í”„)
            fig.add_trace(
                go.Scatter(
                    x=categories,
                    y=prices,
                    mode='lines+markers+text',
                    name=f'{region} í‰ê· ê°€ê²©',
                    line=dict(color='black', width=4),
                    marker=dict(size=12, color='white', line=dict(color='black', width=3)),
                    text=[f'{p:,.0f}ë§Œì›' for p in prices],
                    textposition='middle right',
                    textfont=dict(color='black', size=11),
                    hovertemplate=f'<b>{region} í‰ê· ê°€ê²©</b><br>ë©´ì êµ¬ë¶„: %{{x}}<br>í‰ê· ê°€ê²©: %{{y:,.0f}}ë§Œì›<extra></extra>',
                    showlegend=(row==1)
                ),
                row=row, col=1,
                secondary_y=True
            )
        
        # ì¶• ë¼ë²¨ ì„¤ì •
        for i in range(1, len(self.regions) + 1):
            fig.update_yaxes(title_text="ê±°ë˜ëŸ‰ (ê±´)", secondary_y=False, row=i, col=1)
            fig.update_yaxes(title_text="í‰ê· ê°€ê²© (ë§Œì›)", secondary_y=True, row=i, col=1)
        
        fig.update_layout(
            title='ğŸ“Š ê·œëª¨ë³„ ê±°ë˜ëŸ‰ vs ê°€ê²© ë™ì‹œ ê·¸ë˜í”„',
            title_x=0.5,
            title_font_size=20,
            height=350 * len(self.regions),
            margin=dict(t=120, b=100, l=80, r=120)
        )
        
        fig.show()
    
    def create_area_ratio_pie_charts(self):
        """ì§€ì—­ë³„ ì†Œí˜•Â·ì¤‘í˜•Â·ëŒ€í˜• ë¹„ì¤‘ íŒŒì´ì°¨íŠ¸"""
        print("\n=== ì§€ì—­ë³„ ë©´ì ë³„ ë¹„ì¤‘ íŒŒì´ì°¨íŠ¸ ===")
        
        fig = make_subplots(
            rows=1, cols=len(self.regions),
            subplot_titles=[f'{region}' for region in self.regions],
            specs=[[{"type": "pie"}] * len(self.regions)]
        )
        
        area_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']  # ì†Œí˜•, ì¤‘í˜•, ëŒ€í˜•
        
        for col, region in enumerate(self.regions, 1):
            if region not in self.area_data:
                continue
                
            area_stats = self.area_data[region]['area_stats']
            
            labels = []
            values = []
            colors = []
            
            for i, category in enumerate(['ì†Œí˜•', 'ì¤‘í˜•', 'ëŒ€í˜•']):
                if category in area_stats.index:
                    count = area_stats.loc[category, 'ê±°ë˜ê±´ìˆ˜']
                    labels.append(f'{category}<br>({count:,}ê±´)')
                    values.append(count)
                    colors.append(area_colors[i])
            
            if values:
                fig.add_trace(
                    go.Pie(
                        labels=labels,
                        values=values,
                        name=region,
                        marker_colors=colors,
                        textinfo='label+percent',
                        textposition='auto',
                        hovertemplate='<b>%{label}</b><br>ë¹„ì¤‘: %{percent}<br>ê±°ë˜ëŸ‰: %{value:,}ê±´<extra></extra>'
                    ),
                    row=1, col=col
                )
        
        fig.update_layout(
            title='ğŸ¥§ ì§€ì—­ë³„ ì†Œí˜•Â·ì¤‘í˜•Â·ëŒ€í˜• ë¹„ì¤‘ íŒŒì´ì°¨íŠ¸',
            title_x=0.5,
            title_font_size=20,
            height=500,
            margin=dict(t=100, b=50)
        )
        
        fig.show()
    
    def create_area_summary_table(self):
        """ë©´ì ë³„ ì¢…í•© ìš”ì•½ í…Œì´ë¸”"""
        print("\n=== ë©´ì ë³„ ì¢…í•© ìš”ì•½ í…Œì´ë¸” ===")
        
        table_data = []
        
        for region in self.regions:
            if region not in self.area_data:
                continue
                
            area_stats = self.area_data[region]['area_stats']
            total_count = area_stats['ê±°ë˜ê±´ìˆ˜'].sum()
            
            for category in ['ì†Œí˜•', 'ì¤‘í˜•', 'ëŒ€í˜•']:
                if category in area_stats.index:
                    stats = area_stats.loc[category]
                    ratio = (stats['ê±°ë˜ê±´ìˆ˜'] / total_count) * 100
                    
                    area_range = self.area_categories.get(category, (85, 200))
                    if category == 'ëŒ€í˜•':
                        area_desc = f"{area_range[0]}ã¡ ì´ìƒ"
                    else:
                        area_desc = f"{area_range[0]}~{area_range[1]}ã¡"
                    
                    table_data.append([
                        region,
                        category,
                        area_desc,
                        f"{stats['í‰ê· ë©´ì ']:.1f}ã¡",
                        f"{stats['í‰ê· í‰ìˆ˜']:.1f}í‰",
                        f"{stats['ê±°ë˜ê±´ìˆ˜']:,}ê±´",
                        f"{ratio:.1f}%",
                        f"{stats['í‰ê· ê°€ê²©']:,.0f}ë§Œì›",
                        f"{stats['ì¤‘ìœ„ê°€ê²©']:,.0f}ë§Œì›"
                    ])
        
        if table_data:
            fig = go.Figure(data=[go.Table(
                header=dict(
                    values=['ì§€ì—­', 'ë©´ì êµ¬ë¶„', 'ë©´ì ë²”ìœ„', 'í‰ê· ë©´ì ', 'í‰ê· í‰ìˆ˜', 'ê±°ë˜ê±´ìˆ˜', 'ë¹„ì¤‘', 'í‰ê· ê°€ê²©', 'ì¤‘ìœ„ê°€ê²©'],
                    fill_color='lightblue',
                    align='center',
                    font=dict(size=12, color='black'),
                    height=45
                ),
                cells=dict(
                    values=list(zip(*table_data)),
                    fill_color='white',
                    align='center',
                    font=dict(size=11),
                    height=35
                )
            )])
            
            fig.update_layout(
                title='ğŸ“‹ ì§€ì—­ë³„ ë©´ì ë³„ ì¢…í•© ë¶„ì„í‘œ',
                title_x=0.5,
                title_font_size=16,
                height=500,
                margin=dict(l=20, r=20, t=80, b=20)
            )
            
            fig.show()
    
    def analyze_area_insights(self):
        """ë©´ì ë³„ ì‹œì¥ ì¸ì‚¬ì´íŠ¸"""
        print("\n=== ğŸ” ë©´ì ë³„ ì‹œì¥ ì¸ì‚¬ì´íŠ¸ ===")
        
        for region in self.regions:
            if region not in self.area_data:
                continue
                
            area_stats = self.area_data[region]['area_stats']
            print(f"\nğŸ  [{region}] ë©´ì ë³„ íŠ¹ì„±:")
            
            total_count = area_stats['ê±°ë˜ê±´ìˆ˜'].sum()
            
            # ê°€ì¥ ë§ì´ ê±°ë˜ë˜ëŠ” ë©´ì ëŒ€
            max_volume_category = area_stats['ê±°ë˜ê±´ìˆ˜'].idxmax()
            max_volume_ratio = (area_stats.loc[max_volume_category, 'ê±°ë˜ê±´ìˆ˜'] / total_count) * 100
            
            print(f"  ğŸ“Š ì£¼ë ¥ ë©´ì ëŒ€: {max_volume_category} ({max_volume_ratio:.1f}%)")
            
            # ê°€ì¥ ë¹„ì‹¼ ë©´ì ëŒ€
            max_price_category = area_stats['í‰ê· ê°€ê²©'].idxmax()
            max_price = area_stats.loc[max_price_category, 'í‰ê· ê°€ê²©']
            min_price_category = area_stats['í‰ê· ê°€ê²©'].idxmin()
            min_price = area_stats.loc[min_price_category, 'í‰ê· ê°€ê²©']
            
            price_gap = ((max_price - min_price) / min_price) * 100
            
            print(f"  ğŸ’° ìµœê³ ê°€: {max_price_category} ({max_price:,.0f}ë§Œì›)")
            print(f"  ğŸ’° ìµœì €ê°€: {min_price_category} ({min_price:,.0f}ë§Œì›)")
            print(f"  ğŸ“ˆ ë©´ì ë³„ ê°€ê²© ê²©ì°¨: {price_gap:.1f}%")
            
            # ë©´ì ëŒ€ë³„ ë¹„ì¤‘ ë¶„ì„
            if 'ì†Œí˜•' in area_stats.index and 'ëŒ€í˜•' in area_stats.index:
                small_ratio = (area_stats.loc['ì†Œí˜•', 'ê±°ë˜ê±´ìˆ˜'] / total_count) * 100
                large_ratio = (area_stats.loc['ëŒ€í˜•', 'ê±°ë˜ê±´ìˆ˜'] / total_count) * 100
                
                if small_ratio > 50:
                    print(f"  ğŸ˜ï¸ ì†Œí˜• ì¤‘ì‹¬ ì‹œì¥ (ì†Œí˜• {small_ratio:.1f}%)")
                elif large_ratio > 40:
                    print(f"  ğŸ¢ ëŒ€í˜• ì„ í˜¸ ì‹œì¥ (ëŒ€í˜• {large_ratio:.1f}%)")
                else:
                    print(f"  âš–ï¸ ê· í˜•ì¡íŒ ì‹œì¥ êµ¬ì¡°")
    
    def run_area_analysis(self):
        """ì „ì²´ ë©´ì ë³„ ë¶„ì„ ì‹¤í–‰"""
        print("ğŸ  ì „ìš©ë©´ì ë³„ ê°€ê²© ë° ê±°ë˜ëŸ‰ ë¶„ì„ ì‹œì‘")
        print("=" * 60)
        
        self.load_and_classify_by_area()
        self.create_area_price_trend_chart()
        self.create_area_volume_price_chart() 
        self.create_area_ratio_pie_charts()
        self.create_area_summary_table()
        self.analyze_area_insights()
        
        print(f"\nâœ… ì „ìš©ë©´ì ë³„ ë¶„ì„ ì™„ë£Œ")
        print(f"   ì†Œí˜•/ì¤‘í˜•/ëŒ€í˜•ë³„ ê°€ê²© íŠ¸ë Œë“œì™€ ê±°ë˜ íŒ¨í„´ì„ ì¢…í•© ë¶„ì„í–ˆìŠµë‹ˆë‹¤.")
        
        return self.area_data

# ë¶„ì„ ì‹¤í–‰
area_analyzer = AreaBasedAnalyzer()
area_results = area_analyzer.run_area_analysis()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
from datetime import datetime
import re
import warnings
warnings.filterwarnings('ignore')

class DistrictLevelAnalyzer:
    def __init__(self, data_path='data/'):
        self.data_path = data_path
        self.regions = ['ê´‘ì£¼', 'ë¶€ì‚°', 'ì œì£¼']
        self.district_data = {}
        self.price_per_sqm_data = {}
        
    def load_and_process_district_data(self):
        """ì§€ì—­ë³„ ë°ì´í„° ë¡œë“œ ë° êµ¬Â·ë™ ë‹¨ìœ„ ì „ì²˜ë¦¬"""
        print("=== êµ¬Â·ë™ ë‹¨ìœ„ ê°€ê²© ë¶„ì„ ë°ì´í„° ì²˜ë¦¬ ===")
        
        for region in self.regions:
            print(f"\nğŸ˜ï¸ [{region}] êµ¬Â·ë™ë³„ ë°ì´í„° ì²˜ë¦¬:")
            
            all_data = []
            
            for i in range(1, 6):
                try:
                    filename = f"{region}_ì•„íŒŒíŠ¸_{i}.csv"
                    filepath = f"{self.data_path}{filename}"
                    
                    # íŒŒì¼ ë¡œë“œ
                    for encoding in ['utf-8', 'cp949', 'euc-kr']:
                        try:
                            df = pd.read_csv(filepath, encoding=encoding)
                            
                            # êµ¬Â·ë™ ì •ë³´ì™€ ê°€ê²© ë°ì´í„° ì²˜ë¦¬
                            processed_df = self.process_district_price_data(df, filename, i)
                            
                            if processed_df is not None and len(processed_df) > 0:
                                all_data.append(processed_df)
                                print(f"  âœ… {filename}: {len(processed_df)}ê±´ ì²˜ë¦¬ì™„ë£Œ")
                            
                            break
                        except UnicodeDecodeError:
                            continue
                            
                except Exception as e:
                    print(f"  âŒ {filename}: {e}")
            
            if all_data:
                # ëª¨ë“  ë°ì´í„° í†µí•©
                combined_df = pd.concat(all_data, ignore_index=True)
                
                # êµ¬Â·ë™ë³„ ê°€ê²© ë¶„ì„
                district_analysis = self.analyze_district_prices(combined_df, region)
                self.district_data[region] = district_analysis
                
                print(f"  ğŸ“Š ì´ {len(combined_df):,}ê±´ â†’ êµ¬Â·ë™ë³„ ë¶„ì„ ì™„ë£Œ")
    
    def process_district_price_data(self, df, filename, file_num):
        """êµ¬Â·ë™ ì •ë³´ ë° ê°€ê²© ë°ì´í„° ì²˜ë¦¬"""
        # ì§€ì—­ ê´€ë ¨ ì»¬ëŸ¼ ì°¾ê¸°
        location_keywords = ['ì‹œêµ°êµ¬', 'êµ¬', 'ë™', 'ì§€ì—­', 'ìœ„ì¹˜', 'ì£¼ì†Œ', 'ë²•ì •ë™', 'í–‰ì •ë™']
        location_cols = [col for col in df.columns if any(keyword in col.lower() for keyword in location_keywords)]
        
        # ë‚ ì§œ ì»¬ëŸ¼ ì²˜ë¦¬
        date_cols = [col for col in df.columns if any(keyword in col.lower() 
                    for keyword in ['ë‚ ì§œ', 'date', 'ê³„ì•½', 'ê±°ë˜', 'ë…„ì›”'])]
        
        # ê°€ê²© ì»¬ëŸ¼ ì²˜ë¦¬
        price_keywords = ['ê°€ê²©', 'ê¸ˆì•¡', 'price', 'ê±°ë˜ê¸ˆì•¡', 'ë§¤ë§¤ê°€']
        price_cols = [col for col in df.columns if any(keyword in col.lower() for keyword in price_keywords)]
        
        # ë©´ì  ì»¬ëŸ¼ ì²˜ë¦¬
        area_keywords = ['ë©´ì ', 'area', 'í‰', 'ã¡', 'm2', 'ì „ìš©ë©´ì ']
        area_cols = [col for col in df.columns if any(keyword in col.lower() for keyword in area_keywords)]
        
        print(f"    ğŸ“ ë°œê²¬ëœ ì§€ì—­ ì»¬ëŸ¼: {location_cols}")
        print(f"    ğŸ’° ë°œê²¬ëœ ê°€ê²© ì»¬ëŸ¼: {price_cols}")
        print(f"    ğŸ“ ë°œê²¬ëœ ë©´ì  ì»¬ëŸ¼: {area_cols}")
        
        if not location_cols or not price_cols or not area_cols:
            print(f"    âš ï¸ í•„ìˆ˜ ì»¬ëŸ¼ ë¶€ì¡±")
            return None
        
        # ë‚ ì§œ íŒŒì‹±
        parsed_dates = None
        if date_cols:
            for date_col in date_cols:
                try:
                    formats = ['%Y-%m-%d', '%Y.%m.%d', '%Y/%m/%d', '%Y-%m', '%Y.%m', '%Y%m']
                    
                    for fmt in formats:
                        try:
                            temp_dates = pd.to_datetime(df[date_col], format=fmt, errors='coerce')
                            if temp_dates.notna().sum() / len(temp_dates) > 0.7:
                                parsed_dates = temp_dates
                                break
                        except:
                            continue
                    
                    if parsed_dates is None:
                        parsed_dates = pd.to_datetime(df[date_col], errors='coerce')
                    
                    if parsed_dates.notna().sum() > len(df) * 0.5:
                        break
                except:
                    continue
        
        # ê°€ê²© ë° ë©´ì  ì²˜ë¦¬
        main_price_col = price_cols[0]
        main_area_col = area_cols[0]
        main_location_col = location_cols[0]
        
        try:
            # ê°€ê²© ì²˜ë¦¬
            if df[main_price_col].dtype == 'object':
                price_clean = df[main_price_col].astype(str)
                price_clean = price_clean.str.replace(',', '').str.replace('ë§Œì›', '').str.replace('ì›', '')
                price_clean = price_clean.str.extract('(\d+\.?\d*)')[0]
                cleaned_prices = pd.to_numeric(price_clean, errors='coerce')
            else:
                cleaned_prices = pd.to_numeric(df[main_price_col], errors='coerce')
            
            # ë©´ì  ì²˜ë¦¬
            if df[main_area_col].dtype == 'object':
                area_clean = df[main_area_col].astype(str)
                area_clean = area_clean.str.replace(',', '').str.extract('(\d+\.?\d*)')[0]
                cleaned_areas = pd.to_numeric(area_clean, errors='coerce')
            else:
                cleaned_areas = pd.to_numeric(df[main_area_col], errors='coerce')
            
            # ì§€ì—­ëª… ì •ë¦¬
            cleaned_locations = df[main_location_col].astype(str)
            
        except Exception as e:
            print(f"    âŒ ë°ì´í„° ì²˜ë¦¬ ì‹¤íŒ¨: {e}")
            return None
        
        # ë°ì´í„° í†µí•©
        df['ê±°ë˜ë‚ ì§œ'] = parsed_dates
        df['ê±°ë˜ê°€ê²©'] = cleaned_prices
        df['ì „ìš©ë©´ì '] = cleaned_areas
        df['ì§€ì—­ëª…'] = cleaned_locations
        df['íŒŒì¼ë²ˆí˜¸'] = file_num
        
        # ã¡ë‹¹ ê°€ê²© ê³„ì‚° (ë§Œì› â†’ ì›/ã¡ ë³€í™˜)
        df['ã¡ë‹¹ê°€ê²©'] = (df['ê±°ë˜ê°€ê²©'] * 10000) / df['ì „ìš©ë©´ì ']
        
        # ìœ íš¨í•œ ë°ì´í„°ë§Œ í•„í„°ë§
        valid_data = df[(df['ê±°ë˜ê°€ê²©'].notna()) & 
                       (df['ì „ìš©ë©´ì '].notna()) & 
                       (df['ì§€ì—­ëª…'].notna()) &
                       (df['ê±°ë˜ê°€ê²©'] > 0) & 
                       (df['ê±°ë˜ê°€ê²©'] < 1000000) &
                       (df['ì „ìš©ë©´ì '] > 10) &
                       (df['ì „ìš©ë©´ì '] < 300) &
                       (df['ã¡ë‹¹ê°€ê²©'] > 100000) &  # ìµœì†Œ 10ë§Œì›/ã¡
                       (df['ã¡ë‹¹ê°€ê²©'] < 50000000)]  # ìµœëŒ€ 5ì²œë§Œì›/ã¡
        
        # ì§€ì—­ëª… ì •ë¦¬ (êµ¬Â·ë™ ì¶”ì¶œ)
        valid_data = self.clean_district_names(valid_data)
        
        return valid_data
    
    def clean_district_names(self, df):
        """êµ¬Â·ë™ëª… ì •ë¦¬ ë° í‘œì¤€í™”"""
        def extract_district(location_str):
            if pd.isna(location_str):
                return None
            
            location_str = str(location_str).strip()
            
            # êµ¬ ë‹¨ìœ„ ì¶”ì¶œ (ì˜ˆ: í•´ìš´ëŒ€êµ¬, ì„œêµ¬, ë¶êµ¬ ë“±)
            gu_pattern = r'([ê°€-í£]+êµ¬)'
            gu_match = re.search(gu_pattern, location_str)
            
            # ë™ ë‹¨ìœ„ ì¶”ì¶œ (ì˜ˆ: ì„¼í…€ë™, ìš°ë™, ì¢Œë™ ë“±)
            dong_pattern = r'([ê°€-í£]+ë™)'
            dong_match = re.search(dong_pattern, location_str)
            
            if gu_match and dong_match:
                return f"{gu_match.group(1)} {dong_match.group(1)}"
            elif gu_match:
                return gu_match.group(1)
            elif dong_match:
                return dong_match.group(1)
            else:
                # ê¸°íƒ€ ì§€ì—­ëª… ì²˜ë¦¬
                if len(location_str) > 1:
                    return location_str[:10]  # ë„ˆë¬´ ê¸´ ê²½ìš° ìë¥´ê¸°
                return None
        
        df['ì •ë¦¬ëœì§€ì—­ëª…'] = df['ì§€ì—­ëª…'].apply(extract_district)
        
        # ìœ íš¨í•œ ì§€ì—­ëª…ì´ ìˆëŠ” ë°ì´í„°ë§Œ ìœ ì§€
        df = df[df['ì •ë¦¬ëœì§€ì—­ëª…'].notna()]
        
        return df
    
    def analyze_district_prices(self, df, region):
        """êµ¬Â·ë™ë³„ ê°€ê²© ë¶„ì„"""
        print(f"\n  ğŸ“Š [{region}] êµ¬Â·ë™ë³„ ê°€ê²© ë¶„ì„:")
        
        # êµ¬Â·ë™ë³„ ì§‘ê³„
        district_stats = df.groupby('ì •ë¦¬ëœì§€ì—­ëª…').agg({
            'ê±°ë˜ê°€ê²©': ['count', 'mean', 'median'],
            'ã¡ë‹¹ê°€ê²©': ['mean', 'median', 'std'],
            'ì „ìš©ë©´ì ': 'mean'
        }).round(0)
        
        district_stats.columns = ['ê±°ë˜ê±´ìˆ˜', 'í‰ê· ê±°ë˜ê°€ê²©', 'ì¤‘ìœ„ê±°ë˜ê°€ê²©', 'í‰ê· ã¡ë‹¹ê°€ê²©', 'ì¤‘ìœ„ã¡ë‹¹ê°€ê²©', 'ã¡ë‹¹ê°€ê²©í‘œì¤€í¸ì°¨', 'í‰ê· ë©´ì ']
        
        # ìµœì†Œ ê±°ë˜ê±´ìˆ˜ í•„í„°ë§ (ì‹ ë¢°ì„± í™•ë³´)
        min_transactions = max(10, len(df) // 100)  # ìµœì†Œ 10ê±´ ë˜ëŠ” ì „ì²´ì˜ 1%
        district_stats = district_stats[district_stats['ê±°ë˜ê±´ìˆ˜'] >= min_transactions]
        
        # ã¡ë‹¹ ê°€ê²© ê¸°ì¤€ ì •ë ¬
        district_stats = district_stats.sort_values('í‰ê· ã¡ë‹¹ê°€ê²©', ascending=False)
        
        print(f"    ğŸ“ ë¶„ì„ ê°€ëŠ¥í•œ ì§€ì—­: {len(district_stats)}ê°œ")
        print(f"    ğŸ“Š ìµœì†Œ ê±°ë˜ê±´ìˆ˜ ê¸°ì¤€: {min_transactions}ê±´")
        
        # Top 3 / Bottom 3 ì¶”ì¶œ
        top3 = district_stats.head(3)
        bottom3 = district_stats.tail(3)
        
        print(f"\n    ğŸ† Top 3 ë¹„ì‹¼ ì§€ì—­:")
        for i, (district, row) in enumerate(top3.iterrows(), 1):
            print(f"      {i}. {district}: {row['í‰ê· ã¡ë‹¹ê°€ê²©']:,.0f}ì›/ã¡ ({row['ê±°ë˜ê±´ìˆ˜']}ê±´)")
        
        print(f"\n    ğŸ’° Top 3 ì €ë ´í•œ ì§€ì—­:")
        for i, (district, row) in enumerate(bottom3.iterrows(), 1):
            print(f"      {i}. {district}: {row['í‰ê· ã¡ë‹¹ê°€ê²©']:,.0f}ì›/ã¡ ({row['ê±°ë˜ê±´ìˆ˜']}ê±´)")
        
        return {
            'district_stats': district_stats,
            'top3': top3,
            'bottom3': bottom3,
            'raw_data': df
        }
    
    def create_district_price_ranking_chart(self):
        """ì§€ì—­ë³„ ã¡ë‹¹ ê°€ê²© ìˆœìœ„ ë°”ì°¨íŠ¸"""
        print("\n=== êµ¬Â·ë™ë³„ ã¡ë‹¹ ê°€ê²© ìˆœìœ„ ë°”ì°¨íŠ¸ ===")
        
        fig = make_subplots(
            rows=len(self.regions), cols=1,
            subplot_titles=[f'{region} - êµ¬Â·ë™ë³„ ã¡ë‹¹ ê°€ê²© ìˆœìœ„ (Top 10)' for region in self.regions],
            vertical_spacing=0.15
        )
        
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for row, region in enumerate(self.regions, 1):
            if region not in self.district_data:
                continue
                
            district_stats = self.district_data[region]['district_stats']
            
            # Top 10ë§Œ í‘œì‹œ
            top10 = district_stats.head(10)
            
            fig.add_trace(
                go.Bar(
                    x=top10['í‰ê· ã¡ë‹¹ê°€ê²©'],
                    y=list(range(len(top10))),
                    orientation='h',
                    name=region,
                    marker_color=colors[region],
                    text=[f'{price:,.0f}ì›/ã¡' for price in top10['í‰ê· ã¡ë‹¹ê°€ê²©']],
                    textposition='auto',
                    hovertemplate='<b>%{customdata}</b><br>ã¡ë‹¹ê°€ê²©: %{x:,.0f}ì›/ã¡<br>ê±°ë˜ê±´ìˆ˜: %{meta:,}ê±´<extra></extra>',
                    customdata=top10.index,
                    meta=top10['ê±°ë˜ê±´ìˆ˜'],
                    showlegend=(row==1)
                ),
                row=row, col=1
            )
            
            # yì¶• ë¼ë²¨ì„ ì§€ì—­ëª…ìœ¼ë¡œ ì„¤ì •
            fig.update_yaxes(
                tickmode='array',
                tickvals=list(range(len(top10))),
                ticktext=top10.index,
                row=row, col=1
            )
        
        fig.update_layout(
            title='ğŸ“Š ì§€ì—­ë³„ êµ¬Â·ë™ë³„ ã¡ë‹¹ ê°€ê²© ìˆœìœ„ (Top 10)',
            title_x=0.5,
            title_font_size=20,
            height=400 * len(self.regions),
            margin=dict(t=120, b=100, l=200, r=120)
        )
        
        fig.update_xaxes(title_text="ã¡ë‹¹ ê°€ê²© (ì›/ã¡)")
        
        fig.show()
    
    def create_price_comparison_heatmap(self):
        """ì§€ì—­ë³„ ê°€ê²© ë¹„êµ íˆíŠ¸ë§µ"""
        print("\n=== ì§€ì—­ë³„ ê°€ê²© ë¹„êµ íˆíŠ¸ë§µ ===")
        
        # ê° ì§€ì—­ì˜ ìƒìœ„ 5ê°œ ì§€ì—­ ì¶”ì¶œ
        heatmap_data = []
        
        for region in self.regions:
            if region not in self.district_data:
                continue
                
            district_stats = self.district_data[region]['district_stats']
            top5 = district_stats.head(5)
            
            for district, row in top5.iterrows():
                heatmap_data.append({
                    'ê´‘ì—­ì‹œ': region,
                    'êµ¬ë™': district,
                    'ã¡ë‹¹ê°€ê²©': row['í‰ê· ã¡ë‹¹ê°€ê²©'],
                    'ê±°ë˜ê±´ìˆ˜': row['ê±°ë˜ê±´ìˆ˜'],
                    'ë¼ë²¨': f"{district}<br>{row['í‰ê· ã¡ë‹¹ê°€ê²©']:,.0f}ì›/ã¡"
                })
        
        if heatmap_data:
            heatmap_df = pd.DataFrame(heatmap_data)
            
            # íˆíŠ¸ë§µ ìƒì„±
            fig = go.Figure(data=go.Scatter(
                x=heatmap_df['ê´‘ì—­ì‹œ'],
                y=heatmap_df['êµ¬ë™'],
                mode='markers+text',
                marker=dict(
                    size=heatmap_df['ê±°ë˜ê±´ìˆ˜'] / 50,  # ê±°ë˜ê±´ìˆ˜ì— ë¹„ë¡€í•œ í¬ê¸°
                    color=heatmap_df['ã¡ë‹¹ê°€ê²©'],
                    colorscale='Viridis',
                    showscale=True,
                    colorbar=dict(title="ã¡ë‹¹ ê°€ê²© (ì›/ã¡)")
                ),
                text=heatmap_df['ë¼ë²¨'],
                textposition='middle center',
                textfont=dict(size=10, color='white'),
                hovertemplate='<b>%{y}</b><br>ê´‘ì—­ì‹œ: %{x}<br>ã¡ë‹¹ê°€ê²©: %{marker.color:,.0f}ì›/ã¡<br>ê±°ë˜ê±´ìˆ˜: %{marker.size}ê±´<extra></extra>'
            ))
            
            fig.update_layout(
                title='ğŸ”¥ ì§€ì—­ë³„ í”„ë¦¬ë¯¸ì—„ êµ¬Â·ë™ ê°€ê²© íˆíŠ¸ë§µ (Top 5)',
                title_x=0.5,
                title_font_size=18,
                xaxis_title='ê´‘ì—­ì‹œ',
                yaxis_title='êµ¬Â·ë™',
                height=600,
                margin=dict(t=100, b=50, l=150, r=100)
            )
            
            fig.show()
    
    def create_top_bottom_comparison(self):
        """ìµœê³ ê°€ vs ìµœì €ê°€ ì§€ì—­ ë¹„êµ"""
        print("\n=== ìµœê³ ê°€ vs ìµœì €ê°€ ì§€ì—­ ë¹„êµ ===")
        
        fig = make_subplots(
            rows=1, cols=2,
            subplot_titles=('ğŸ† ì§€ì—­ë³„ ìµœê³ ê°€ êµ¬Â·ë™', 'ğŸ’° ì§€ì—­ë³„ ìµœì €ê°€ êµ¬Â·ë™'),
            specs=[[{"type": "bar"}, {"type": "bar"}]]
        )
        
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        # ìµœê³ ê°€ ì§€ì—­
        top_districts = []
        bottom_districts = []
        
        for region in self.regions:
            if region not in self.district_data:
                continue
                
            top3 = self.district_data[region]['top3']
            bottom3 = self.district_data[region]['bottom3']
            
            if not top3.empty:
                top_district = top3.iloc[0]
                top_districts.append({
                    'region': region,
                    'district': top_district.name,
                    'price': top_district['í‰ê· ã¡ë‹¹ê°€ê²©'],
                    'count': top_district['ê±°ë˜ê±´ìˆ˜']
                })
            
            if not bottom3.empty:
                bottom_district = bottom3.iloc[-1]
                bottom_districts.append({
                    'region': region,
                    'district': bottom_district.name,
                    'price': bottom_district['í‰ê· ã¡ë‹¹ê°€ê²©'],
                    'count': bottom_district['ê±°ë˜ê±´ìˆ˜']
                })
        
        # ìµœê³ ê°€ ì°¨íŠ¸
        if top_districts:
            top_df = pd.DataFrame(top_districts)
            fig.add_trace(
                go.Bar(
                    x=top_df['region'],
                    y=top_df['price'],
                    name='ìµœê³ ê°€',
                    marker_color=[colors[region] for region in top_df['region']],
                    text=[f"{district}<br>{price:,.0f}ì›/ã¡" for district, price in zip(top_df['district'], top_df['price'])],
                    textposition='auto',
                    hovertemplate='<b>%{x}</b><br>ì§€ì—­: %{customdata}<br>ã¡ë‹¹ê°€ê²©: %{y:,.0f}ì›/ã¡<extra></extra>',
                    customdata=top_df['district']
                ),
                row=1, col=1
            )
        
        # ìµœì €ê°€ ì°¨íŠ¸
        if bottom_districts:
            bottom_df = pd.DataFrame(bottom_districts)
            fig.add_trace(
                go.Bar(
                    x=bottom_df['region'],
                    y=bottom_df['price'],
                    name='ìµœì €ê°€',
                    marker_color=[colors[region] for region in bottom_df['region']],
                    text=[f"{district}<br>{price:,.0f}ì›/ã¡" for district, price in zip(bottom_df['district'], bottom_df['price'])],
                    textposition='auto',
                    hovertemplate='<b>%{x}</b><br>ì§€ì—­: %{customdata}<br>ã¡ë‹¹ê°€ê²©: %{y:,.0f}ì›/ã¡<extra></extra>',
                    customdata=bottom_df['district'],
                    showlegend=False
                ),
                row=1, col=2
            )
        
        fig.update_layout(
            title='âš–ï¸ ì§€ì—­ë³„ ìµœê³ ê°€ vs ìµœì €ê°€ êµ¬Â·ë™ ë¹„êµ',
            title_x=0.5,
            title_font_size=18,
            height=500,
            margin=dict(t=100, b=50)
        )
        
        fig.update_yaxes(title_text="ã¡ë‹¹ ê°€ê²© (ì›/ã¡)")
        
        fig.show()
    
    def create_district_summary_table(self):
        """êµ¬Â·ë™ë³„ ì¢…í•© ìš”ì•½ í…Œì´ë¸”"""
        print("\n=== êµ¬Â·ë™ë³„ ì¢…í•© ìš”ì•½ í…Œì´ë¸” ===")
        
        table_data = []
        
        for region in self.regions:
            if region not in self.district_data:
                continue
                
            # ìƒìœ„ 5ê°œ, í•˜ìœ„ 3ê°œ ì§€ì—­ í¬í•¨
            top5 = self.district_data[region]['district_stats'].head(5)
            bottom3 = self.district_data[region]['district_stats'].tail(3)
            
            combined = pd.concat([top5, bottom3])
            
            for district, row in combined.iterrows():
                price_gap = ((row['í‰ê· ã¡ë‹¹ê°€ê²©'] - self.district_data[region]['district_stats']['í‰ê· ã¡ë‹¹ê°€ê²©'].min()) / 
                           self.district_data[region]['district_stats']['í‰ê· ã¡ë‹¹ê°€ê²©'].min()) * 100
                
                table_data.append([
                    region,
                    district,
                    f"{row['ê±°ë˜ê±´ìˆ˜']:,}ê±´",
                    f"{row['í‰ê· ã¡ë‹¹ê°€ê²©']:,.0f}ì›/ã¡",
                    f"{row['ì¤‘ìœ„ã¡ë‹¹ê°€ê²©']:,.0f}ì›/ã¡", 
                    f"{row['í‰ê· ê±°ë˜ê°€ê²©']:,.0f}ë§Œì›",
                    f"{row['í‰ê· ë©´ì ']:.1f}ã¡",
                    f"{price_gap:.1f}%"
                ])
        
        if table_data:
            fig = go.Figure(data=[go.Table(
                header=dict(
                    values=['ê´‘ì—­ì‹œ', 'êµ¬Â·ë™', 'ê±°ë˜ê±´ìˆ˜', 'í‰ê· ã¡ë‹¹ê°€ê²©', 'ì¤‘ìœ„ã¡ë‹¹ê°€ê²©', 'í‰ê· ê±°ë˜ê°€ê²©', 'í‰ê· ë©´ì ', 'ìµœì €ê°€ëŒ€ë¹„'],
                    fill_color='lightblue',
                    align='center',
                    font=dict(size=12, color='black'),
                    height=45
                ),
                cells=dict(
                    values=list(zip(*table_data)),
                    fill_color='white',
                    align='center',
                    font=dict(size=11),
                    height=35
                )
            )])
            
            fig.update_layout(
                title='ğŸ“‹ êµ¬Â·ë™ë³„ ë¶€ë™ì‚° ê°€ê²© ì¢…í•© ë¶„ì„í‘œ',
                title_x=0.5,
                title_font_size=16,
                height=600,
                margin=dict(l=20, r=20, t=80, b=20)
            )
            
            fig.show()
    
    def analyze_district_insights(self):
        """êµ¬Â·ë™ë³„ ì‹œì¥ ì¸ì‚¬ì´íŠ¸"""
        print("\n=== ğŸ” êµ¬Â·ë™ë³„ ì‹œì¥ ì¸ì‚¬ì´íŠ¸ ===")
        
        for region in self.regions:
            if region not in self.district_data:
                continue
                
            district_stats = self.district_data[region]['district_stats']
            print(f"\nğŸ˜ï¸ [{region}] êµ¬Â·ë™ë³„ íŠ¹ì„±:")
            
            # ê°€ê²© ê²©ì°¨ ë¶„ì„
            max_price = district_stats['í‰ê· ã¡ë‹¹ê°€ê²©'].max()
            min_price = district_stats['í‰ê· ã¡ë‹¹ê°€ê²©'].min()
            price_gap = ((max_price - min_price) / min_price) * 100
            
            print(f"  ğŸ“Š ìµœê³ ê°€ vs ìµœì €ê°€ ê²©ì°¨: {price_gap:.1f}%")
            print(f"    ìµœê³ : {district_stats['í‰ê· ã¡ë‹¹ê°€ê²©'].idxmax()} ({max_price:,.0f}ì›/ã¡)")
            print(f"    ìµœì €: {district_stats['í‰ê· ã¡ë‹¹ê°€ê²©'].idxmin()} ({min_price:,.0f}ì›/ã¡)")
            
            # ê±°ë˜ í™œë°œë„ ë¶„ì„
            total_transactions = district_stats['ê±°ë˜ê±´ìˆ˜'].sum()
            avg_transactions = district_stats['ê±°ë˜ê±´ìˆ˜'].mean()
            
            most_active = district_stats['ê±°ë˜ê±´ìˆ˜'].idxmax()
            most_active_count = district_stats.loc[most_active, 'ê±°ë˜ê±´ìˆ˜']
            
            print(f"  ğŸ“ˆ ê°€ì¥ í™œë°œí•œ ì§€ì—­: {most_active} ({most_active_count:,}ê±´)")
            print(f"  ğŸ“Š í‰ê·  ê±°ë˜ëŸ‰: {avg_transactions:.0f}ê±´/ì§€ì—­")
            
            # í”„ë¦¬ë¯¸ì—„ ì§€ì—­ íŠ¹ì„±
            premium_threshold = district_stats['í‰ê· ã¡ë‹¹ê°€ê²©'].quantile(0.8)  # ìƒìœ„ 20%
            premium_districts = district_stats[district_stats['í‰ê· ã¡ë‹¹ê°€ê²©'] >= premium_threshold]
            
            print(f"  ğŸ† í”„ë¦¬ë¯¸ì—„ ì§€ì—­ ({len(premium_districts)}ê°œ): ã¡ë‹¹ {premium_threshold:,.0f}ì› ì´ìƒ")
            
            # ê°€ì„±ë¹„ ì§€ì—­ ì¶”ì²œ
            value_districts = district_stats[
                (district_stats['í‰ê· ã¡ë‹¹ê°€ê²©'] <= district_stats['í‰ê· ã¡ë‹¹ê°€ê²©'].median()) &
                (district_stats['ê±°ë˜ê±´ìˆ˜'] >= district_stats['ê±°ë˜ê±´ìˆ˜'].median())
            ]
            
            if not value_districts.empty:
                print(f"  ğŸ’ ê°€ì„±ë¹„ ì¶”ì²œ ì§€ì—­: {', '.join(value_districts.head(3).index)}")
    
    def run_district_analysis(self):
        """ì „ì²´ êµ¬Â·ë™ë³„ ë¶„ì„ ì‹¤í–‰"""
        print("ğŸ˜ï¸ êµ¬Â·ë™ë³„ ê°€ê²© ë¶„ì„ ì‹œì‘")
        print("=" * 60)
        
        self.load_and_process_district_data()
        self.create_district_price_ranking_chart()
        self.create_price_comparison_heatmap()
        self.create_top_bottom_comparison()
        self.create_district_summary_table()
        self.analyze_district_insights()
        
        print(f"\nâœ… êµ¬Â·ë™ë³„ ê°€ê²© ë¶„ì„ ì™„ë£Œ")
        print(f"   ê° ì§€ì—­ì˜ í•«í”Œë ˆì´ìŠ¤ì™€ ê°€ì„±ë¹„ ì¢‹ì€ ì§€ì—­ì„ ë¶„ì„í–ˆìŠµë‹ˆë‹¤.")
        
        return self.district_data

# ë¶„ì„ ì‹¤í–‰
district_analyzer = DistrictLevelAnalyzer()
district_results = district_analyzer.run_district_analysis()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
from datetime import datetime, timedelta
from scipy import stats
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.stattools import grangercausalitytests
import warnings
warnings.filterwarnings('ignore')

class TimeLagMarketAnalyzer:
    def __init__(self, data_path='data/'):
        self.data_path = data_path
        self.regions = ['ê´‘ì£¼', 'ë¶€ì‚°', 'ì œì£¼']
        self.monthly_panel_data = {}
        self.lag_analysis_results = {}
        self.market_indicators = {}
        
    def create_monthly_panel_data(self):
        """ì›”ë³„ íŒ¨ë„ë°ì´í„° ìƒì„± (YYYY-MM í˜•íƒœ)"""
        print("=== 1ï¸âƒ£ ì›”ë³„ íŒ¨ë„ë°ì´í„° ìƒì„± ===")
        
        for region in self.regions:
            print(f"\nğŸ“Š [{region}] ì›”ë³„ ì§‘ê³„:")
            
            all_data = []
            
            for i in range(1, 6):
                try:
                    filename = f"{region}_ì•„íŒŒíŠ¸_{i}.csv"
                    filepath = f"{self.data_path}{filename}"
                    
                    # íŒŒì¼ ë¡œë“œ
                    for encoding in ['utf-8', 'cp949', 'euc-kr']:
                        try:
                            df = pd.read_csv(filepath, encoding=encoding)
                            
                            # ë‚ ì§œ, ê°€ê²© ë°ì´í„° ì²˜ë¦¬
                            processed_df = self.process_monthly_data(df, filename, i)
                            
                            if processed_df is not None and len(processed_df) > 0:
                                all_data.append(processed_df)
                                print(f"  âœ… {filename}: {len(processed_df)}ê±´ ì²˜ë¦¬ì™„ë£Œ")
                            
                            break
                        except UnicodeDecodeError:
                            continue
                            
                except Exception as e:
                    print(f"  âŒ {filename}: {e}")
            
            if all_data:
                # ëª¨ë“  ë°ì´í„° í†µí•©
                combined_df = pd.concat(all_data, ignore_index=True)
                
                # ì›”ë³„ ì§‘ê³„
                monthly_agg = self.aggregate_monthly_data(combined_df, region)
                self.monthly_panel_data[region] = monthly_agg
                
                print(f"  ğŸ“… ì›”ë³„ ì§‘ê³„ ì™„ë£Œ: {len(monthly_agg)}ê°œì›”")
    
    def process_monthly_data(self, df, filename, file_num):
        """ì›”ë³„ ë°ì´í„° ì²˜ë¦¬"""
        # ë‚ ì§œ ì»¬ëŸ¼ ì²˜ë¦¬
        date_cols = [col for col in df.columns if any(keyword in col.lower() 
                    for keyword in ['ë‚ ì§œ', 'date', 'ê³„ì•½', 'ê±°ë˜', 'ë…„ì›”'])]
        
        # ê°€ê²© ì»¬ëŸ¼ ì²˜ë¦¬
        price_keywords = ['ê°€ê²©', 'ê¸ˆì•¡', 'price', 'ê±°ë˜ê¸ˆì•¡', 'ë§¤ë§¤ê°€']
        price_cols = [col for col in df.columns if any(keyword in col.lower() for keyword in price_keywords)]
        
        if not date_cols or not price_cols:
            return None
        
        # ë‚ ì§œ íŒŒì‹±
        parsed_dates = None
        for date_col in date_cols:
            try:
                formats = ['%Y-%m-%d', '%Y.%m.%d', '%Y/%m/%d', '%Y-%m', '%Y.%m', '%Y%m']
                
                for fmt in formats:
                    try:
                        temp_dates = pd.to_datetime(df[date_col], format=fmt, errors='coerce')
                        if temp_dates.notna().sum() / len(temp_dates) > 0.7:
                            parsed_dates = temp_dates
                            break
                    except:
                        continue
                
                if parsed_dates is None:
                    parsed_dates = pd.to_datetime(df[date_col], errors='coerce')
                
                if parsed_dates.notna().sum() > len(df) * 0.5:
                    break
            except:
                continue
        
        # ê°€ê²© íŒŒì‹±
        main_price_col = price_cols[0]
        try:
            if df[main_price_col].dtype == 'object':
                price_clean = df[main_price_col].astype(str)
                price_clean = price_clean.str.replace(',', '').str.replace('ë§Œì›', '').str.replace('ì›', '')
                price_clean = price_clean.str.extract('(\d+\.?\d*)')[0]
                cleaned_prices = pd.to_numeric(price_clean, errors='coerce')
            else:
                cleaned_prices = pd.to_numeric(df[main_price_col], errors='coerce')
        except:
            return None
        
        # ë°ì´í„° í†µí•©
        df['ê±°ë˜ë‚ ì§œ'] = parsed_dates
        df['ê±°ë˜ê°€ê²©'] = cleaned_prices
        
        # ìœ íš¨í•œ ë°ì´í„°ë§Œ í•„í„°ë§
        valid_data = df[(df['ê±°ë˜ë‚ ì§œ'].notna()) & 
                       (df['ê±°ë˜ê°€ê²©'].notna()) & 
                       (df['ê±°ë˜ê°€ê²©'] > 0) & 
                       (df['ê±°ë˜ê°€ê²©'] < 1000000)]
        
        return valid_data
    
    def aggregate_monthly_data(self, df, region):
        """ì›”ë³„ ì§‘ê³„"""
        # ë…„ì›” ì»¬ëŸ¼ ìƒì„±
        df['ë…„ì›”'] = df['ê±°ë˜ë‚ ì§œ'].dt.to_period('M')
        
        # ì›”ë³„ ì§‘ê³„
        monthly_agg = df.groupby('ë…„ì›”').agg({
            'ê±°ë˜ê°€ê²©': ['count', 'mean', 'median'],
            'ê±°ë˜ë‚ ì§œ': ['min', 'max']
        }).round(2)
        
        monthly_agg.columns = ['transaction_count', 'price_avg', 'price_median', 'period_start', 'period_end']
        
        # í•„ìˆ˜ ì»¬ëŸ¼ ìƒì„±
        monthly_agg = monthly_agg.reset_index()
        monthly_agg['date'] = monthly_agg['ë…„ì›”'].astype(str)
        monthly_agg['region'] = region
        monthly_agg['year_month'] = pd.to_datetime(monthly_agg['date'])
        
        # ì‹œê³„ì—´ ìˆœì„œ ì •ë ¬
        monthly_agg = monthly_agg.sort_values('year_month')
        
        return monthly_agg
    
    def exploratory_data_analysis(self):
        """2ï¸âƒ£ íƒìƒ‰ì  ë¶„ì„ (EDA)"""
        print("\n=== 2ï¸âƒ£ íƒìƒ‰ì  ë¶„ì„ (EDA) ===")
        
        # 1. ê±°ë˜ëŸ‰ê³¼ ê°€ê²© ì¶”ì„¸ ì‹œê°í™”
        self.create_dual_axis_timeseries()
        
        # 2. ê¸°ë³¸ ìƒê´€ê´€ê³„ ë¶„ì„
        self.analyze_basic_correlation()
    
    def create_dual_axis_timeseries(self):
        """ê±°ë˜ëŸ‰ê³¼ ê°€ê²© ì¶”ì„¸ ì‹œê°í™” (ì´ì¤‘ì¶•)"""
        print("\nğŸ“ˆ ê±°ë˜ëŸ‰-ê°€ê²© ì‹œê³„ì—´ ì‹œê°í™”:")
        
        fig = make_subplots(
            rows=len(self.regions), cols=1,
            subplot_titles=[f'{region} - ê±°ë˜ëŸ‰ vs ê°€ê²© ì¶”ì„¸ (ì„ í–‰-í›„í–‰ ê´€ê³„ í™•ì¸)' for region in self.regions],
            specs=[[{"secondary_y": True}] for _ in self.regions],
            vertical_spacing=0.15
        )
        
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for row, region in enumerate(self.regions, 1):
            if region in self.monthly_panel_data:
                data = self.monthly_panel_data[region]
                
                # ê±°ë˜ëŸ‰ (ë§‰ëŒ€ê·¸ë˜í”„)
                fig.add_trace(
                    go.Bar(
                        x=data['year_month'],
                        y=data['transaction_count'],
                        name=f'{region} ê±°ë˜ëŸ‰',
                        marker_color=colors[region],
                        opacity=0.6,
                        hovertemplate=f'<b>{region} ê±°ë˜ëŸ‰</b><br>ë‚ ì§œ: %{{x}}<br>ê±°ë˜ëŸ‰: %{{y:,}}ê±´<extra></extra>',
                        showlegend=(row==1)
                    ),
                    row=row, col=1,
                    secondary_y=False
                )
                
                # í‰ê· ê°€ê²© (ì„ ê·¸ë˜í”„)
                fig.add_trace(
                    go.Scatter(
                        x=data['year_month'],
                        y=data['price_avg'],
                        mode='lines+markers',
                        name=f'{region} í‰ê· ê°€ê²©',
                        line=dict(color=colors[region], width=3),
                        marker=dict(size=6),
                        hovertemplate=f'<b>{region} í‰ê· ê°€ê²©</b><br>ë‚ ì§œ: %{{x}}<br>í‰ê· ê°€ê²©: %{{y:,.0f}}ë§Œì›<extra></extra>',
                        showlegend=(row==1)
                    ),
                    row=row, col=1,
                    secondary_y=True
                )
        
        # ì¶• ë¼ë²¨ ì„¤ì •
        for i in range(1, len(self.regions) + 1):
            fig.update_yaxes(title_text="ê±°ë˜ëŸ‰ (ê±´)", secondary_y=False, row=i, col=1)
            fig.update_yaxes(title_text="í‰ê· ê°€ê²© (ë§Œì›)", secondary_y=True, row=i, col=1)
        
        fig.update_layout(
            title='ğŸ“Š ê±°ë˜ëŸ‰-ê°€ê²© ì‹œê³„ì—´ ë¶„ì„ (ì„ í–‰-í›„í–‰ ê´€ê³„ í™•ì¸)',
            title_x=0.5,
            title_font_size=20,
            height=350 * len(self.regions),
            margin=dict(t=120, b=100, l=80, r=120)
        )
        
        fig.show()
    
    def analyze_basic_correlation(self):
        """ê¸°ë³¸ ìƒê´€ê´€ê³„ ë¶„ì„"""
        print("\nğŸ” ê¸°ë³¸ ìƒê´€ê´€ê³„ ë¶„ì„:")
        
        for region in self.regions:
            if region not in self.monthly_panel_data:
                continue
                
            data = self.monthly_panel_data[region]
            
            if len(data) > 5:
                corr_coef = data['transaction_count'].corr(data['price_avg'])
                print(f"  ğŸ“Š {region}: ê±°ë˜ëŸ‰-ê°€ê²© ìƒê´€ê³„ìˆ˜ = {corr_coef:.3f}")
                
                if corr_coef > 0.3:
                    print(f"    â†’ ì–‘ì˜ ìƒê´€ê´€ê³„ (ê±°ë˜ëŸ‰â†‘ â†’ ê°€ê²©â†‘)")
                elif corr_coef < -0.3:
                    print(f"    â†’ ìŒì˜ ìƒê´€ê´€ê³„ (ê±°ë˜ëŸ‰â†‘ â†’ ê°€ê²©â†“)")
                else:
                    print(f"    â†’ ì•½í•œ ìƒê´€ê´€ê³„")
    
    def time_lag_analysis(self):
        """3ï¸âƒ£ ì‹œì°¨ ë¶„ì„ (Time-lagged Analysis)"""
        print("\n=== 3ï¸âƒ£ ì‹œì°¨ ë¶„ì„ (Time-lagged Analysis) ===")
        
        # A. ë‹¨ìˆœ ìƒê´€ê³„ìˆ˜ (Lag Correlation)
        self.lag_correlation_analysis()
        
        # B. íšŒê·€ ë¶„ì„ (Distributed Lag Model)
        self.distributed_lag_model()
        
        # C. ê·¸ëœì € ì¸ê³¼ ê²€ì • (ì„ íƒì  - ë°ì´í„° ì¶©ë¶„ì‹œ)
        self.granger_causality_test()
    
    def lag_correlation_analysis(self):
        """ë°©ë²• A: ë‹¨ìˆœ ìƒê´€ê³„ìˆ˜ (Lag Correlation)"""
        print("\nğŸ“Š ë°©ë²• A: ì‹œì°¨ ìƒê´€ê³„ìˆ˜ ë¶„ì„")
        
        lag_results = {}
        
        for region in self.regions:
            if region not in self.monthly_panel_data:
                continue
                
            data = self.monthly_panel_data[region].copy()
            
            if len(data) < 10:  # ìµœì†Œ 10ê°œì›” ë°ì´í„° í•„ìš”
                continue
            
            print(f"\n  ğŸ  [{region}] ì‹œì°¨ë³„ ìƒê´€ê³„ìˆ˜:")
            
            lag_correlations = []
            
            # 1ê°œì›”ë¶€í„° 6ê°œì›”ê¹Œì§€ ì‹œì°¨ ë¶„ì„
            for lag in range(1, 7):
                if len(data) > lag:
                    # ê±°ë˜ëŸ‰ì„ lagë§Œí¼ ì•ìœ¼ë¡œ ë‹¹ê²¨ì„œ ë¯¸ë˜ ê°€ê²©ê³¼ ë¹„êµ
                    volume_lagged = data['transaction_count'].iloc[:-lag]
                    price_future = data['price_avg'].iloc[lag:]
                    
                    if len(volume_lagged) > 3:
                        corr_coef = volume_lagged.corr(price_future)
                        lag_correlations.append({
                            'lag_months': lag,
                            'correlation': corr_coef,
                            'sample_size': len(volume_lagged)
                        })
                        
                        significance = "ğŸ”¥ ë†’ìŒ" if abs(corr_coef) > 0.5 else "ğŸ“ˆ ë³´í†µ" if abs(corr_coef) > 0.3 else "ğŸ“Š ë‚®ìŒ"
                        print(f"    {lag}ê°œì›” í›„ ê°€ê²© ì˜ˆì¸¡ë ¥: {corr_coef:+.3f} ({significance})")
            
            lag_results[region] = lag_correlations
        
        self.lag_analysis_results['correlations'] = lag_results
        
        # ì‹œì°¨ ìƒê´€ê³„ìˆ˜ ì‹œê°í™”
        self.visualize_lag_correlations(lag_results)
    
    def visualize_lag_correlations(self, lag_results):
        """ì‹œì°¨ ìƒê´€ê³„ìˆ˜ ì‹œê°í™”"""
        fig = go.Figure()
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for region, correlations in lag_results.items():
            if correlations:
                lags = [item['lag_months'] for item in correlations]
                corrs = [item['correlation'] for item in correlations]
                
                fig.add_trace(go.Scatter(
                    x=lags,
                    y=corrs,
                    mode='lines+markers',
                    name=f'{region}',
                    line=dict(color=colors[region], width=3),
                    marker=dict(size=8),
                    hovertemplate=f'<b>{region}</b><br>ì‹œì°¨: %{{x}}ê°œì›”<br>ìƒê´€ê³„ìˆ˜: %{{y:.3f}}<extra></extra>'
                ))
        
        fig.add_hline(y=0, line=dict(color='gray', dash='dash', width=1))
        fig.add_hline(y=0.3, line=dict(color='green', dash='dot', width=1), annotation_text="ë³´í†µ ìƒê´€ê´€ê³„")
        fig.add_hline(y=-0.3, line=dict(color='red', dash='dot', width=1))
        
        fig.update_layout(
            title='ğŸ“ˆ ê±°ë˜ëŸ‰ â†’ ê°€ê²© ì‹œì°¨ë³„ ìƒê´€ê³„ìˆ˜ (ì˜ˆì¸¡ë ¥ ë¶„ì„)',
            title_x=0.5,
            title_font_size=18,
            xaxis_title='ì‹œì°¨ (ê°œì›”)',
            yaxis_title='ìƒê´€ê³„ìˆ˜',
            height=500,
            xaxis=dict(dtick=1)
        )
        
        fig.show()
    
    def distributed_lag_model(self):
        """ë°©ë²• B: ë¶„ì‚°ì§€ì—°ëª¨í˜• (Distributed Lag Model)"""
        print("\nğŸ“Š ë°©ë²• B: ë¶„ì‚°ì§€ì—°ëª¨í˜• íšŒê·€ë¶„ì„")
        
        regression_results = {}
        
        for region in self.regions:
            if region not in self.monthly_panel_data:
                continue
                
            data = self.monthly_panel_data[region].copy()
            
            if len(data) < 15:  # ìµœì†Œ 15ê°œì›” ë°ì´í„° í•„ìš”
                continue
            
            print(f"\n  ğŸ  [{region}] íšŒê·€ë¶„ì„ ê²°ê³¼:")
            
            # ë…ë¦½ë³€ìˆ˜: 1~3ê°œì›” ì „ ê±°ë˜ëŸ‰
            # ì¢…ì†ë³€ìˆ˜: í˜„ì¬ ê°€ê²©
            max_lag = min(3, len(data) - 5)
            
            if max_lag < 1:
                continue
            
            # ë°ì´í„° ì¤€ë¹„
            y = data['price_avg'].iloc[max_lag:].values  # í˜„ì¬ ê°€ê²©
            X = []
            
            for lag in range(1, max_lag + 1):
                volume_lag = data['transaction_count'].iloc[max_lag-lag:-lag].values
                X.append(volume_lag)
            
            X = np.column_stack(X)
            
            if len(X) > 5:
                # ì„ í˜•íšŒê·€ ì‹¤í–‰
                reg = LinearRegression()
                reg.fit(X, y)
                
                r_squared = reg.score(X, y)
                
                print(f"    ğŸ“Š ê²°ì •ê³„ìˆ˜ (RÂ²): {r_squared:.3f}")
                print(f"    ğŸ“ˆ íšŒê·€ê³„ìˆ˜:")
                
                for i, coef in enumerate(reg.coef_, 1):
                    significance = "ğŸ”¥" if abs(coef) > 50 else "ğŸ“ˆ" if abs(coef) > 20 else "ğŸ“Š"
                    print(f"      {i}ê°œì›” ì „ ê±°ë˜ëŸ‰: {coef:+.2f} {significance}")
                
                regression_results[region] = {
                    'r_squared': r_squared,
                    'coefficients': reg.coef_,
                    'intercept': reg.intercept_
                }
        
        self.lag_analysis_results['regression'] = regression_results
    
    def granger_causality_test(self):
        """ë°©ë²• C: ê·¸ëœì € ì¸ê³¼ê²€ì • (ì„ íƒì )"""
        print("\nğŸ“Š ë°©ë²• C: ê·¸ëœì € ì¸ê³¼ê²€ì •")
        
        granger_results = {}
        
        for region in self.regions:
            if region not in self.monthly_panel_data:
                continue
                
            data = self.monthly_panel_data[region].copy()
            
            if len(data) < 20:  # ìµœì†Œ 20ê°œì›” ë°ì´í„° í•„ìš”
                print(f"  âš ï¸ {region}: ë°ì´í„° ë¶€ì¡± ({len(data)}ê°œì›”)")
                continue
            
            try:
                # ë°ì´í„° ì¤€ë¹„ (ê²°ì¸¡ê°’ ì œê±°)
                test_data = data[['transaction_count', 'price_avg']].dropna()
                
                if len(test_data) < 15:
                    continue
                
                # ê·¸ëœì € ì¸ê³¼ê²€ì • ì‹¤í–‰ (ìµœëŒ€ 6ê°œì›” ì‹œì°¨)
                max_lag = min(6, len(test_data) // 4)
                
                result = grangercausalitytests(test_data[['price_avg', 'transaction_count']], 
                                             maxlag=max_lag, verbose=False)
                
                print(f"  ğŸ  [{region}] ê·¸ëœì € ì¸ê³¼ê²€ì •:")
                
                significant_lags = []
                for lag in range(1, max_lag + 1):
                    p_value = result[lag][0]['ssr_ftest'][1]  # F-test p-value
                    
                    if p_value < 0.05:
                        significant_lags.append(lag)
                        print(f"    {lag}ê°œì›”: p-value = {p_value:.3f} âœ… ìœ ì˜ë¯¸")
                    else:
                        print(f"    {lag}ê°œì›”: p-value = {p_value:.3f}")
                
                if significant_lags:
                    print(f"    ğŸ¯ ê±°ë˜ëŸ‰ì´ ê°€ê²©ì„ ì˜ˆì¸¡í•˜ëŠ” ìµœì  ì‹œì°¨: {significant_lags}ê°œì›”")
                else:
                    print(f"    ğŸ“Š í†µê³„ì ìœ¼ë¡œ ìœ ì˜ë¯¸í•œ ì¸ê³¼ê´€ê³„ ì—†ìŒ")
                
                granger_results[region] = {
                    'significant_lags': significant_lags,
                    'test_results': result
                }
                
            except Exception as e:
                print(f"  âŒ {region}: ê·¸ëœì € ê²€ì • ì‹¤íŒ¨ - {e}")
        
        self.lag_analysis_results['granger'] = granger_results
    
    def market_turning_point_indicators(self):
        """4ï¸âƒ£ ì‹œì¥ ì „í™˜ì  ì˜ˆì¸¡ ì§€í‘œ ê°œë°œ"""
        print("\n=== 4ï¸âƒ£ ì‹œì¥ ì „í™˜ì  ì˜ˆì¸¡ ì§€í‘œ ê°œë°œ ===")
        
        for region in self.regions:
            if region not in self.monthly_panel_data:
                continue
                
            data = self.monthly_panel_data[region].copy()
            
            if len(data) < 12:
                continue
            
            print(f"\nğŸ¯ [{region}] ì‹œì¥ ì „í™˜ì  ì§€í‘œ:")
            
            # 1. Volume-to-Price Ratio (VPR)
            data['VPR'] = data['transaction_count'] / (data['price_avg'] / 1000)  # ì²œë§Œì› ë‹¨ìœ„ë¡œ ì •ê·œí™”
            
            # 2. ê±°ë˜ëŸ‰ ì´ë™í‰ê·  êµì°¨ ì§€í‘œ
            data['volume_ma_3'] = data['transaction_count'].rolling(window=3).mean()
            data['volume_ma_12'] = data['transaction_count'].rolling(window=12).mean()
            data['volume_cross_signal'] = data['volume_ma_3'] / data['volume_ma_12']
            
            # 3. ê±°ë˜ëŸ‰ ê¸‰ì¦ë¥  + ê°€ê²© ìƒìŠ¹ë¥  ì¡°í•© ì§€í‘œ
            data['volume_change_3m'] = data['transaction_count'].pct_change(3) * 100
            data['price_change_3m'] = data['price_avg'].pct_change(3) * 100
            data['combined_momentum'] = data['volume_change_3m'] + data['price_change_3m']
            
            # ìµœê·¼ 3ê°œì›” ì§€í‘œ ì¶œë ¥
            recent_data = data.tail(3)
            
            print(f"  ğŸ“Š ìµœê·¼ 3ê°œì›” ì§€í‘œ ë³€í™”:")
            for _, row in recent_data.iterrows():
                date_str = row['date']
                vpr = row['VPR'] if pd.notna(row['VPR']) else 0
                cross_signal = row['volume_cross_signal'] if pd.notna(row['volume_cross_signal']) else 1
                momentum = row['combined_momentum'] if pd.notna(row['combined_momentum']) else 0
                
                # ì‹ í˜¸ í•´ì„
                vpr_signal = "ğŸ”¥ ê¸‰ë“± ì‹ í˜¸" if vpr > data['VPR'].quantile(0.8) else "â„ï¸ ì¹¨ì²´ ì‹ í˜¸" if vpr < data['VPR'].quantile(0.2) else "ğŸ“Š ë³´í†µ"
                cross_signal_text = "ğŸ“ˆ í™œí™©" if cross_signal > 1.1 else "ğŸ“‰ ì¹¨ì²´" if cross_signal < 0.9 else "â¡ï¸ ë³´í•©"
                momentum_signal = "ğŸš€ ìƒìŠ¹" if momentum > 20 else "ğŸ”» í•˜ë½" if momentum < -20 else "â¡ï¸ íš¡ë³´"
                
                print(f"    {date_str}: VPR={vpr:.2f} ({vpr_signal}), êµì°¨={cross_signal:.2f} ({cross_signal_text}), ëª¨ë©˜í…€={momentum:+.1f}% ({momentum_signal})")
            
            self.market_indicators[region] = data
        
        # ì§€í‘œ ì‹œê°í™”
        self.visualize_market_indicators()
    
    def visualize_market_indicators(self):
        """ì‹œì¥ ì „í™˜ì  ì§€í‘œ ì‹œê°í™”"""
        print("\nğŸ“ˆ ì‹œì¥ ì „í™˜ì  ì§€í‘œ ì‹œê°í™”")
        
        fig = make_subplots(
            rows=3, cols=1,
            subplot_titles=('VPR (Volume-to-Price Ratio)', 'ê±°ë˜ëŸ‰ ì´ë™í‰ê·  êµì°¨', 'ê²°í•© ëª¨ë©˜í…€ ì§€í‘œ'),
            vertical_spacing=0.1
        )
        
        colors = {'ê´‘ì£¼': '#FF6B6B', 'ë¶€ì‚°': '#4ECDC4', 'ì œì£¼': '#45B7D1'}
        
        for region in self.regions:
            if region not in self.market_indicators:
                continue
                
            data = self.market_indicators[region]
            
            # 1. VPR
            fig.add_trace(
                go.Scatter(
                    x=data['year_month'],
                    y=data['VPR'],
                    mode='lines',
                    name=f'{region} VPR',
                    line=dict(color=colors[region], width=2),
                    showlegend=True
                ),
                row=1, col=1
            )
            
            # 2. ê±°ë˜ëŸ‰ êµì°¨ ì‹ í˜¸
            fig.add_trace(
                go.Scatter(
                    x=data['year_month'],
                    y=data['volume_cross_signal'],
                    mode='lines',
                    name=f'{region} êµì°¨ì‹ í˜¸',
                    line=dict(color=colors[region], width=2),
                    showlegend=False
                ),
                row=2, col=1
            )
            
            # 3. ê²°í•© ëª¨ë©˜í…€
            fig.add_trace(
                go.Scatter(
                    x=data['year_month'],
                    y=data['combined_momentum'],
                    mode='lines',
                    name=f'{region} ëª¨ë©˜í…€',
                    line=dict(color=colors[region], width=2),
                    showlegend=False
                ),
                row=3, col=1
            )
        
        # ê¸°ì¤€ì„  ì¶”ê°€
        fig.add_hline(y=1, line=dict(color='gray', dash='dash', width=1), row=2, col=1)
        fig.add_hline(y=0, line=dict(color='gray', dash='dash', width=1), row=3, col=1)
        
        fig.update_layout(
            title='ğŸ¯ ì‹œì¥ ì „í™˜ì  ì˜ˆì¸¡ ì§€í‘œ ëŒ€ì‹œë³´ë“œ',
            title_x=0.5,
            title_font_size=20,
            height=800,
            margin=dict(t=120, b=100)
        )
        
        fig.show()
    
    def generate_market_forecast_summary(self):
        """ì‹œì¥ ì „ë§ ìš”ì•½"""
        print("\n=== ğŸ”® ì‹œì¥ ì „ë§ ìš”ì•½ ===")
        
        for region in self.regions:
            if region not in self.market_indicators:
                continue
                
            data = self.market_indicators[region]
            recent = data.tail(1).iloc[0]
            
            print(f"\nğŸ  [{region}] í˜„ì¬ ì‹œì¥ ìƒí™©:")
            
            # ì¢…í•© ì‹ í˜¸ íŒë‹¨
            signals = []
            
            # VPR ì‹ í˜¸
            vpr_current = recent['VPR'] if pd.notna(recent['VPR']) else 0
            vpr_threshold_high = data['VPR'].quantile(0.75)
            vpr_threshold_low = data['VPR'].quantile(0.25)
            
            if vpr_current > vpr_threshold_high:
                signals.append("ìƒìŠ¹")
                vpr_signal = "ğŸ”¥ ê³¼ì—´ (ë§¤ë„ ì‹ í˜¸)"
            elif vpr_current < vpr_threshold_low:
                signals.append("í•˜ë½")
                vpr_signal = "â„ï¸ ì¹¨ì²´ (ë§¤ìˆ˜ ì‹ í˜¸)"
            else:
                signals.append("ë³´í•©")
                vpr_signal = "ğŸ“Š ì ì • ìˆ˜ì¤€"
            
            # êµì°¨ ì‹ í˜¸
            cross_current = recent['volume_cross_signal'] if pd.notna(recent['volume_cross_signal']) else 1
            if cross_current > 1.1:
                signals.append("ìƒìŠ¹")
                cross_signal = "ğŸ“ˆ í™œí™© ì „í™˜"
            elif cross_current < 0.9:
                signals.append("í•˜ë½")
                cross_signal = "ğŸ“‰ ì¹¨ì²´ ì „í™˜"
            else:
                signals.append("ë³´í•©")
                cross_signal = "â¡ï¸ í˜„ìƒ ìœ ì§€"
            
            # ëª¨ë©˜í…€ ì‹ í˜¸
            momentum_current = recent['combined_momentum'] if pd.notna(recent['combined_momentum']) else 0
            if momentum_current > 20:
                signals.append("ìƒìŠ¹")
                momentum_signal = "ğŸš€ ê°•í•œ ìƒìŠ¹ ëª¨ë©˜í…€"
            elif momentum_current < -20:
                signals.append("í•˜ë½")
                momentum_signal = "ğŸ”» ê°•í•œ í•˜ë½ ëª¨ë©˜í…€"
            else:
                signals.append("ë³´í•©")
                momentum_signal = "â¡ï¸ ì•½í•œ ëª¨ë©˜í…€"
            
            # ì¢…í•© íŒë‹¨
            signal_count = {"ìƒìŠ¹": signals.count("ìƒìŠ¹"), "í•˜ë½": signals.count("í•˜ë½"), "ë³´í•©": signals.count("ë³´í•©")}
            dominant_signal = max(signal_count, key=signal_count.get)
            
            print(f"  ğŸ’¡ VPR ì§€í‘œ: {vpr_signal}")
            print(f"  ğŸ’¡ êµì°¨ ì§€í‘œ: {cross_signal}")
            print(f"  ğŸ’¡ ëª¨ë©˜í…€ ì§€í‘œ: {momentum_signal}")
            
            if dominant_signal == "ìƒìŠ¹":
                overall_forecast = "ğŸš€ ìƒìŠ¹ ì „ë§"
                recommendation = "ë§¤ìˆ˜ ê²€í†  ì‹œì "
            elif dominant_signal == "í•˜ë½":
                overall_forecast = "ğŸ“‰ í•˜ë½ ì „ë§"
                recommendation = "ë§¤ë„ ê²€í†  ë˜ëŠ” ê´€ë§"
            else:
                overall_forecast = "â¡ï¸ íš¡ë³´ ì „ë§"
                recommendation = "í˜„ìƒ ìœ ì§€"
            
            print(f"  ğŸ¯ ì¢…í•© ì „ë§: {overall_forecast}")
            print(f"  ğŸ“‹ íˆ¬ì ë°©í–¥: {recommendation}")
            
            # ì‹ ë¢°ë„ í‰ê°€
            signal_strength = max(signal_count.values()) / len(signals)
            if signal_strength >= 0.67:
                confidence = "ë†’ìŒ ğŸ”¥"
            elif signal_strength >= 0.5:
                confidence = "ë³´í†µ ğŸ“Š"
            else:
                confidence = "ë‚®ìŒ âš ï¸"
            
            print(f"  ğŸ“Š ì‹ ë¢°ë„: {confidence} ({signal_strength:.1%})")
    
    def create_comprehensive_summary_table(self):
        """ì¢…í•© ë¶„ì„ ìš”ì•½ í…Œì´ë¸”"""
        print("\n=== ğŸ“‹ ì¢…í•© ë¶„ì„ ìš”ì•½ í…Œì´ë¸” ===")
        
        table_data = []
        
        for region in self.regions:
            if region not in self.monthly_panel_data:
                continue
            
            # ê¸°ë³¸ í†µê³„
            data = self.monthly_panel_data[region]
            avg_volume = data['transaction_count'].mean()
            avg_price = data['price_avg'].mean()
            
            # ì‹œì°¨ ë¶„ì„ ê²°ê³¼
            best_lag = "N/A"
            best_correlation = 0
            
            if region in self.lag_analysis_results.get('correlations', {}):
                correlations = self.lag_analysis_results['correlations'][region]
                if correlations:
                    best_item = max(correlations, key=lambda x: abs(x['correlation']))
                    best_lag = f"{best_item['lag_months']}ê°œì›”"
                    best_correlation = best_item['correlation']
            
            # í˜„ì¬ ì§€í‘œ ê°’
            current_vpr = "N/A"
            current_momentum = "N/A"
            
            if region in self.market_indicators:
                recent = self.market_indicators[region].tail(1).iloc[0]
                current_vpr = f"{recent['VPR']:.2f}" if pd.notna(recent['VPR']) else "N/A"
                current_momentum = f"{recent['combined_momentum']:+.1f}%" if pd.notna(recent['combined_momentum']) else "N/A"
            
            table_data.append([
                region,
                f"{len(data)}ê°œì›”",
                f"{avg_volume:,.0f}ê±´",
                f"{avg_price:,.0f}ë§Œì›",
                best_lag,
                f"{best_correlation:+.3f}",
                current_vpr,
                current_momentum
            ])
        
        if table_data:
            fig = go.Figure(data=[go.Table(
                header=dict(
                    values=['ì§€ì—­', 'ë¶„ì„ê¸°ê°„', 'í‰ê· ê±°ë˜ëŸ‰', 'í‰ê· ê°€ê²©', 'ìµœì ì‹œì°¨', 'ìµœê³ ìƒê´€ê³„ìˆ˜', 'í˜„ì¬VPR', 'í˜„ì¬ëª¨ë©˜í…€'],
                    fill_color='lightblue',
                    align='center',
                    font=dict(size=12, color='black'),
                    height=45
                ),
                cells=dict(
                    values=list(zip(*table_data)),
                    fill_color='white',
                    align='center',
                    font=dict(size=11),
                    height=35
                )
            )])
            
            fig.update_layout(
                title='ğŸ“Š ê±°ë˜ëŸ‰-ê°€ê²© ì‹œì°¨ë¶„ì„ ë° ì‹œì¥ì§€í‘œ ì¢…í•©ìš”ì•½',
                title_x=0.5,
                title_font_size=16,
                height=400,
                margin=dict(l=20, r=20, t=80, b=20)
            )
            
            fig.show()
    
    def run_comprehensive_analysis(self):
        """ì „ì²´ ë¶„ì„ ì‹¤í–‰"""
        print("ğŸ”¬ ê±°ë˜ëŸ‰-ê°€ê²© ì‹œì°¨ë¶„ì„ ë° ì‹œì¥ ì „í™˜ì  ì˜ˆì¸¡ ë¶„ì„ ì‹œì‘")
        print("=" * 80)
        
        # 1ë‹¨ê³„: ì›”ë³„ íŒ¨ë„ë°ì´í„° ìƒì„±
        self.create_monthly_panel_data()
        
        # 2ë‹¨ê³„: íƒìƒ‰ì  ë¶„ì„
        self.exploratory_data_analysis()
        
        # 3ë‹¨ê³„: ì‹œì°¨ ë¶„ì„
        self.time_lag_analysis()
        
        # 4ë‹¨ê³„: ì‹œì¥ ì „í™˜ì  ì§€í‘œ
        self.market_turning_point_indicators()
        
        # 5ë‹¨ê³„: ì¢…í•© ì „ë§
        self.generate_market_forecast_summary()
        
        # 6ë‹¨ê³„: ìš”ì•½ í…Œì´ë¸”
        self.create_comprehensive_summary_table()
        
        print(f"\nâœ… ê³ ê¸‰ ì‹œê³„ì—´ ë¶„ì„ ì™„ë£Œ")
        print(f"   ê±°ë˜ëŸ‰ì˜ ê°€ê²© ì˜ˆì¸¡ë ¥ê³¼ ì‹œì¥ ì „í™˜ì ì„ ì¢…í•© ë¶„ì„í–ˆìŠµë‹ˆë‹¤.")
        print(f"   ğŸ¯ í•µì‹¬ ë°œê²¬: ê±°ë˜ëŸ‰ì´ ê°€ê²©ì„ ì„ í–‰í•˜ëŠ” ìµœì  ì‹œì°¨ì™€ í˜„ì¬ ì‹œì¥ ì‹ í˜¸")
        
        return {
            'monthly_data': self.monthly_panel_data,
            'lag_analysis': self.lag_analysis_results,
            'market_indicators': self.market_indicators
        }

# ë¶„ì„ ì‹¤í–‰
time_lag_analyzer = TimeLagMarketAnalyzer()
comprehensive_results = time_lag_analyzer.run_comprehensive_analysis()
```
